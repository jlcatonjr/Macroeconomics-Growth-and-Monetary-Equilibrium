In a moneyed economy, money is half of every transaction. It is money that links activity across the economic system, facilitating the rationalization of production toward the end of efficiently serving the effective demands of consumers. Without money, the scope of economic organization must necessarily be inhibited by high transaction costs. It is the ability of money to reduce these costs that drive exchanging agents to use money as a medium of exchange.

The widespread use of money in an economy gives it special status within macroeconomic analysis. Perturbations in money markets ripple outward into the the market for goods. Changes in the value of production impact money markets, as do changes in consumer and investor preference to hold money. Money markets and goods markets are inextricably linked as money is half of every exchange. The health of one effects the health of the other.

In this chapter, we will:

* consider the factors surrounding money's emergence.
* relate this emergence to the forces of supply and demand.
* elaborate the meaning of the price of money.
* consider the role of the state in money production.

To understand the significance of money, it is helpful to discuss how incentives likely drove the emergence of money. As you might expect, we will use are standard supply and demand analysis to build this narrative. The move from primitive barter to indirect exchange is a story of unintended consequences. Economic theory has much to offer in clarifying these consequences in the form of monetary theory.

Much of this discussion centers on the role of the value of money in guiding production. We orient our understanding using the equation of exchange. The average price of money, defined as the inverse of the price level, allows monetary dynamics to be expalined in terms of the equation of exchange. The quantity of money also influences its own value, and by implication, the price level. Along with the interest rate - the rental price of money - we will consider in chapter 8 how the value of money influences its quantity in the forms of base money and credit.

* The Evolution of Money

To the average observer, the existence of money is given. Money is a good, but unlike other goods, its primary use is as a medium of exchange [1]. In the pre-modern era, money was typically linked to a commodity, but in the modern financial system, this is no longer true – at least not for base money provided by central banks. So how can money become an object of its own, delinked from any use value? The modern money is the result of a long process of evolution. We begin by describing the mechanisms that allowed for the earliest development of money.

Individuals only act if they expect that action to promote an improvement in value compared to other actions. Every agent has opportunities to engage in acts of entrepreneurship intended to create value. She imagines that she can transform the world from its present state into one that she prefers more greatly. She may prove successful and attain a profit, meaning that she enters a state that she values more greatly than the one she would experience otherwise. She may break even in terms of change in net value experience, or even incur a loss. 

The concept extends to exchange. Two agents, both looking to improve the state of their existence, notice that each has a good desired by the other. It so happens that each wants the good that the other holds so they exchange the goods. Each has improved his or her lot, although we cannot be sure by how much exactly as there is no such thing as a cardinal utility measure that enables objective comparison of agent valuations. We must interpret agent action at face value and accept the action as a contextually constrained expression of an agent's preference to realize the greatest net value .

Barter is easy to accommodate in a model where trading agents lack geography. Action, however, always occurs at a particular time and place. The specificity of time an place implies a cost for an agent who seeks to find an appropriate trading partner. The agent interested in a trade, let’s call him agent A, must find another, agent B, who owns the desired object and who is interested in trading it for something owned by agent A. Often, this double coincidence of wants fails to arise between two bartering agents. 

The agent can continue looking for a single trading partner, or he can partition the task. Instead of attaining the desired good in a single exchange, he can find a good that is demanded by agent B and trade that intermediate good for good he desires. Over time, the agent might realize that there are one or several goods that most easily accommodate this indirect exchange. Others may observe that this practice is a wise one and begin to copy the innovation. Goods that are recognized as being helpful in accommodating exchange are adopted for the purpose of exchange. Those most useful for this purpose become money: a commonly accepted medium of exchange.

Money does not arise by the plan of a single individual. It may initially emerge without explicit intention that the good become universally accepted within the exchange economy. The goal of agent A was simply to find a good desired by agent B. There is no need for agent A to expect that others will adopt his strategy. As the innovation is copied, the commodity becomes a network good. It gains value because other agents are willing to use it in indirect exchange. As exchanging parties increasingly expect that the intermediate good will be accepted by others, they can charge prices in terms of that good. The good that becomes money consequently serves as a common numeraire in which prices of non-money goods are denominated.

# The Qualities and Functions of Money

What makes for good money? Societies have tended to adopt money that meets 5 criteria. Money must be:

* durable
* easily divisible
* relatively scarce
* highly saleable
* portable

These qualities promote the function of money. With these criteria more or less present, money can serve as:

* a medium of exchange
* a store of value
* a unit of account
* a standard of deferred payment

Money provides its primary service as a commonly accepted medium of exchange. This role allows for and is supported by other monetary functions. Notice that different criteria for money relate to different functions of money. 

1. Saleability and portability allow money to function as a medium of exchange in the first place. How could a good act as a medium of exchange if it is difficult to transport or if nobody wants to accept it? 
2. Durability and scarcity promote money’s function as a store of value. Money that cannot maintain its value, either due to deterioration or due to ease of production or attainment, cannot be used to defer consumption or investment to a later date.
3. Divisibility is closely linked to money’s role as a unit of account. Prices, denominated in the unit of account, are more easily accommodated if money can be divided into small, homogeneous units. A good's weight often served this function in the world of commodity money. 
4. Given a common unit of account, agents can also lend money to one another. This allows an agent to attain a good that she otherwise could not afford or would be unable to borrow. Thus, money becomes a standard of deferred payment.

Most significant for economic rationality, money represents an innovation in accounting. Even without double entry book keeping, monetary units themselves serve as a physical system of accounting. The giving up of a unit of money necessarily prevents that unit from being spent again elsewhere.

A common standard of measure enables the approximation of a good's value compared to all other goods. Since prices are denominated in a common unit of account, the value of different goods, as determined in light of the demand of consumers and cost of supply, can be compared objectively. These prices represent the quantity of money for which a good has been sold in the market.

A common unit of account promotes the rationalization of production. Producers to compare costs and revenues in the same monetary unit so that they can be certain of the magnitudes of their gain or loss. Profits calculated form market prices may  be used to rank the efficiency of different producers of goods, with those least efficient producers earning negative profits.  Prices also allow consumers to more efficiently compare the cost incurred in their pursuit of value, thus increasing the net value derived from a  given level of income.

## The Price of Money

Just as markets select for a good that can efficiently serve as a medium of exchange, the quantity of money is also determined by market forces. Imagine that in a market dependent on barter exchange, people realize that oats can be used as a commonly excepted medium of exchange. As more people realize this, oats are increasingly demanded for their ability to aid exchange. This change means that any person who buys or sells goods is willing to accept oats in exchange. A farmer who sells an apple now negotiates a price in terms of some weight or volume of oats. He buys other goods using these same oats. This increase in demand for oats increases the price that a unit of oats commands in the market.

If oats are money, how do we determine their price. For any money, its price is what it exchanges for. If a quart of oats trades for 2 apples in equilibrium, then the price of a quart of oats is 2 apples. If the same quart trades for 3 oranges, then the price of a quart of oats is also 3 oranges. The price of an apple is 1/2 quart of oats and the price of an orange is 1/3 quart of oats.

Money, then, has an array of prices. The price of money is whatever it can exchange for. What we usually mean by the price of money, however, is the average price of money. This is often referred to as its purchasing power of money $P_M$. This refers to the inverse of the price level, $P$. A price index considers all exchanges for each type of good. The price of each category of good sold is weighted according to the total number of goods exchanged.

In reality, the basket of goods and weights of these goods change overtime, so there exists no perfect estimation of the price level and, by implication, the purchasing power of money. Still, it is useful to attempt to estimate changes in the average price of money as changes in the supply and demand for money can lead to changes in demand for for many other goods. We will see in the following chapters how changes in the supply and demand for money are often responsible for economic fluctuations.

Money has at least as many prices as there are goods that it can exchange for. Even for transactions of the same kind of good, each transaction may generate a unique price. Competitive markets tend to promote convergence toward a common price for each good. This is most apparent in financial markets where the cost of buying and selling goods in different markets is especially small. Rarely do price discrepancies persist between different exchanges. Despite the existence of unique prices in particular exchanges, financial markets anchor the prices of goods and tend to reflect the law of one price. This operates both for the price of non-money goods and for the price of one money relative to another. This tendency justifies the use of a single price of money in analysis, where this price is the inverse of the price level.

The price of money is supported by two sources of demand for money: use value and exchange value. In order for money to evolve from a world defined by barter relations, the good that is to be adopted as money must have some naturally occurring use. In the example of oats, the good has use in consumption by humans or may be used as feed for animals. Value from these original uses of oats comprise the use value of oats. Before oats are used as money, use value is represented by the equilibrium price of oats, $P_{O_0}$. Suppose that oats are recognized as being useful as a medium of exchange. With a new source of value, demand for oats increases. If the quantity of oats is fixed, then the exchange value of oats is represented by the increase in price, $P_{O_1}-P_{O_0}$. Below we consider a market where the supply of oats is upward sloping.

A farmer who sells an apple can negotiate a price in terms of some weight of oats. He buys other goods using these same oats. Originally oats have value because they can be consumed. This original value makes oats saleable even before they are used as an intermediate good to accommodate exchange. If individuals realize that oats accommodate exchange, this new function adds a new source of demand in the oats market. Increase demand for oats will lead to an increase in the price of oats. This increase in price, before quantity has adjusted, reflects the exchange value of oats.

After the quantity supplied has adjusted, the use value of oats is reflected by the price that that is implied by the intersection of the dotted vertical line representing $Q_1$ and the original demand curve, $D_0$. The exchange value of oats is represented by the distance between $D_0$ and $D_1$ at the quantity $Q_1$. In the short-run, the price must adjust to reflect changes in demand for a good and its factors of supply. In the long-run, quantity adjusts, alleviating part of the burden placed on prices.

The higher price sustained by the use of oats as medium of exchange incentivizes greater production of oats. If the price is pushed high enough, it can also incentivize the development of technology that reduces the cost of oat production. Something similar occurred at the end of the 19th century when a relatively high price of gold incentivized the development of cyanide extraction process. For oats, this may include the development of new irrigation techniques or systematic use and breeding of the most productive strains of oats, among other possibilities. If the technology successfully reduces the cost of producing a unit of oats, the cost of supply falls. We represent this by shifting the supply curve to the right (Figure 3). Suppose that oat farmers suffer an ongoing drought. What will happen to the price of oats? In the case of a drought, the cost of supply increases. The supply curve for oats would shift left and the value of oats would increase. With a sustained, high level of demand, the rise in price may also serve as impetus to develop technology that increases yields.

## Money and the Equation of Exchange

Money plays a significant role in the equation of exchange as its existence and use define aggregate demand- i.e., total expenditures within a specific time period for an economy. It is an essential component of every exchange in a monetary economy. We can therefore link changes to the money market to changes in the market for non-money goods. 

Recall that the equation of exchange can, which in its standard form represents aggregate demand aggregate supply, can be transformed to represent the supply of and demand for money:

>$MV\equiv Py$
>
>$M\equiv\frac{Py}{V}$

Replace $V$ with $\frac{1}{k}$

>$M\equiv Pyk$
>
>$M_S\equiv M_D$

At any moment, the quantity of money in circulation must be demanded for transactions, $Py$, or held as part of portfolio demand for money, $k$, representing money held throughout the period relative to the value of total expenditures. Money held throughout the period cannot be used for expenditure and vice versa. 

This does not quite reflect the whole situation, however. An increase in portfolio demand money may influence either $P$ or $y$, depending on the circumstance. Likewise, as the economy becomes more productive, reflected by an increase in $y$, it is also true that the price level must fall absent an increase in the willingness of money holders to spend a greater portion of their cash balances.

To account for this nuance, we define money demanded in terms of real cash balances by adjusting the value of money for changes in the price level.

>$\frac{M}{P}\equiv yk$
>
>$m_s\equiv m_D$

The demand for  real cash balances must match the value of the real stock of money.  If either $y$ or $k$ increase, the real stock of money must increase by an increase in $M$ or decrease in $P$. This is particularly important for cases where $k$ increases. Absent an adjustment in the real money stock, a rise in $k$ can initiate a fall in $y$ in the short run. In such a situation, the economy fails to reach a monetary equilibrium where the economy is producing real income at its maximum sustainable level. 

If demand to hold money remains constant, a fall in $P$ results in an increase in real income, $y$, often referred to as a wealth effect. In the next chapter, we will return to this idea in thinking about the role of a fall in the price level when the total value of expenditures has fallen due to an increase in $k$.

## Price Dynamics of the Money Market

An increase in real income yields an increase in real cash balances. This occurs either through a drop in the price level that results or an increase in $M$ representing the markets response to an increase in demand to spend money. We will consider the first case here and the latter case after aggregate demand is introduced. In the modern environment, the quantity of base money is governed by central bank policy. Policy is ultimately discretionary, meaning that the central bank does not have to  adjust the quantity of money in light of changes in money's price. Ex ante, we cannot be sure what policy the monetary authorities will pursue. This setting is therefore represented with a perfectly inelastic supply curve.
Consideration of corollary changes in the money and goods markets are linked by aggregated demand and the relationship of the average price of money to the price level. Since the price level is a measure of average prices, the average price of money is the inverse of the price level:

> $P_M=\frac{1}{P}$
  
If the price of money falls, the price level rises and vice versa. In the next chapter, we will consider in detail how changes in the money market to changes in the goods market. For now, we will consider factors that might influence the price of money.

The price of money will respond to either a change in the demand for or supply of money. A change in demand for money could affect the quantity of money supplied if the supply curve is not perfectly inelastic. In this case, the change in the price of money be offset in the long-run by an increase in the quantity of money. If the money stock is perfectly inelastic, then the change in price will be the long-run effect of a change in demand.

The price of money increases if either the supply of money decreases or the demand for money increases. Since the price of money is the inverse of the price level, the price level will decrease in this case. This change in demand for money can come from one of two sources. There can be an increase in portfolio demand for money, $k$. This is demand to hold money. Or there can be an increase in transactions demand for money, $y$. This is demand to spend money.

It should come as no surprise that the price of money will fall if either the supply of money increases or demand for money falls. Again, this change in demand can come in the form of a decrease in transactions demand for money, $y$, or a decrease in portfolio demand for money, $k$.

To sum, the relationship between changes in the money market and changes in the price of money and in the price level are as follows:

| | + | - |
|--|--|--|
|Supply of Money |$\downarrow P_M $|$\uparrow P_M$|
|Demand for Money |$\uparrow P_M $|$\downarrow P_M$| 

If the money under consideration is produced by the  market, its quantity will change alongside changes in the purchasing power of money. With changes in the supply of money, this much is obvious. Changes in demand are positively correlated with changes in the money stock. As the demand for money increases, a rise in the price of money will lead producers of money to provide more to the market. Likewise, a fall in the price of money driven by a fall in demand will reduce the rate of money creation by money producers.

## A State Monopoly over Money


For much of recorded human history, the state, meaning the government, has been involved in the production of money. The earliest metal coins were found in Lydia, a small kingdom in what is now Turkey. The faces of rulers are found on these coins.

The process that is most likely responsible for the emergence of money as a concept has already been described. Once money enters widespread use, new opportunities arise to transform the monetary unity. As society converged to gold, silver, and copper for facilitating commerce, this presented an opportunity for the ruling class and those aspiring to comprise it to exercise control over the monetary system. The role of money minting commands an appreciable amount of power in a nascent market with few competitors. As in any market, this power increases with the establishment and enforcement of monopoly privilege by the state. 

There is no a priori reason that we should expect a monopoly on money to occur. David Glasner argues that, for the production of a good to be a natural monopoly, the technology must exhibit economies of scale that ensure that the average cost of production is always lower if one produces the entire output of an industry than if two or more firms with access to the identical technology divide the output [1]. But even if the state were the lowest-cost producer of money, it would not necessarily enjoy the economies of scale required for the existence of a natural monopoly. 

Of prime concern in the ancient world was fall in a currency's value owing to manipulation of the monetary unit. Such manipulation would alter the composition of the money stock in circulation. When devalued money of a particular nominal value circulate alongside coins of the same denomination but of different, more highly valued composition, there arises a an increase in the costs of exchange. Agents engaged in exchange must expend energy to ascertain a coin’s metallic content. 

Imagine that we have two coins that weigh 1 oz and have different compositions. Coin A is 4 parts silver and 1 part gold (or in other words, 4/5 of an oz silver and 1/5 oz gold). Coin B is 4 parts gold, 1 part silver. Assuming that gold is worth more than silver, coin B is worth more than coin A. Since it is convenient to trade with coins of the same denomination, agents will continue to trade with coins of this 1 oz denomination, but they will tend to remove the coins with more gold from circulation. 

These coins serve as a store of value – defense against inflation – while the cheaper money is employed as a medium of exchange. Of course, coin A might also serve as a store of value, but the chance of continued devaluation makes ownership of the coin A more attractive for this purpose. This outcome is typified by Gresham’s law.

The sovereign often printed his seal on money produced in his realm. The seal of the sovereign may have given “traders more confidence in the weight and fineness of coins.” The sovereign, however, was “seeking not to improve the monetary system but only to exploit profit opportunity implicit in this premium”. If a public good was provided, it was likely an outcome of the sovereign's pursuit of one's own own interest. The seal of the sovereign increased confidence in both the value of the coin and the stability of the political regime.

The core of the explanation must consider the motives of agents close to power and involved in coining. Glasner offers one clear explanation for the emergence of state monopoly: defense. High-minded principles may guide the sovereign attempt to stabilize and improve the monetary system, but they would not explain incentive for doing so. That is, moralistic narratives do not represent an economic explanation for the near universal tendency for the sovereign to monopolize production of money. Competition for power is fierce. Any leader, elected or unelected, faces competition from numerous others who would prefer to have the power for themselves and their own allies. One regime is always at risk of being displaced by another. A weapon that an agent or team of agents might potentially use is influence over the quantity of money. 

Minting a large quantity of debased coins might enable a private mint owner to finance an attempt to overthrow an incumbent sovereign. To be sure, such a debasement would violate the mint owner’s promises about the content of the coins he was issuing. But upon becoming the sovereign, the owner could avoid any legal liability by annulling his legal obligation to those he had defrauded. A mint-master who sought to overthrow the existing regime could inhibit money’s role as a unit of account by debauching it while increasing his own wealth in the process.  Given this threat, monopolization of the production of money is not only an opportunity for a sovereign to profit, but also to keep political competitors at bay. A sovereign might well feel justified, then, in not only controlling the mint, but also establishing legal tender monopoly. 

This sort of problem is reflected in the writing of Nicolas Copernicus, most famous for his work in astronomy. In his “On the Minting of Money”, he argues:

>It would be advantageous therefore for there to be only one common mint for all Prussia, in which every type of money would be stamped on one side with the insignia of the lands of Prussia: they should have a crown at the top, so that the superiority of the kingdom would be recognized. On the obverse, the insignia of the duke of Prussia could be seen under the crown above it.

Copernicus was worried about mint-masters in Prussia who threatened the stability of the Prussian empire by minting debauched currency. Surely Copernicus was familiar with the political instability that might arise from debasement.

That a mint-master could successfully destabilize the ruling regime before going out of business was another concern. Hume seems to have had something similar in mind, though his worry concerned the extension of credit by private banks, as opposed to dishonest coinage. Not only did he suggest that a single bank should regulate the credit stock – presumably the Bank of England – he even suggested that it is the job of the sovereign to ensure its modest increase: 

>The good policy of the magistrate consists only in keeping it, if possible, still increasing; because, by that means, he keeps alive a spirit of an industry in the nation and increases the stock of labour, in which consists all real power and riches.

Oddly enough, the end of Hume’s life coincides with the early years of relative stability under the free banking regime in Scotland. His insight is correct that instability in the economy often leads to political instability. Hume’s emphasis on internal stability and progress fits well with Glasner's theory.     

The process also works in the reverse direction. Control over the money supply provides a source of funds during an emergency. Defense is important both to the internal politic and in foreign affairs.  Even in a regime that typically promotes policies supporting sound money, major wars tend inevitably to devalue the currency. This source may be abused during periods of instability, but a sovereign who acts with restraint could have avoided such abuse.  

When a nation enters into war, the monetary system becomes a tool by which resources can be coordinated toward the war effort. This was a recurring theme in the ancient world. We also witness this in modern and early modern wars. Whether one considers the depreciation of the continental during the American Revolution or that of the British pound during the Napoleonic wars or the widespread suspension of the gold standard during World War I, modern governments have a habit of funding wartime expenditures via monetary inflation much like their counterparts in antiquity. 

Having considered the emergence of a state monopoly over money, the analysis has been primarily concerned with premodern and early modern institutions. This leaves us to consider whether or not a state monopoly over the production of money - at least base money - is still justified. The predominant modern monetary doctrine suggests that state control over money is an important tool for maintaining macroeconomic stability. We will consider the prospect of macroeconomic stability using theory before returning to the issue of state involvement in management of the  monetary system. 

# Conclusion

This chapter has introduced money's role, emergence, and the forces determining its demand and supply. It is unlikely that money, as a concept, arose by any  means except trial and error among exchanging parties. Standard supply and demand analysis suggest that, whatever commodity was selected as money, the market would provide more of the commodity in response to an increase in demand for the it. By this means, the quantity of money in circulation adjusts to demand for it.
We also considered the role of the state in the development of monetary economies. The pervasiveness of money - that it is included in one side of every exchange in a monetary economy - made state involvement in the production and regulation of money both a profitable enterprise and an attractive defensive strategy.
