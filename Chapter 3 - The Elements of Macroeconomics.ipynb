Imagine that you are explaining the game of basketball to someone who has never seen the sport before. To explain the game you need to identify basic objects used in the game: players, the court and the boundaries it includes, a basketball, and a basket. You would have to explain how the objects are related, or in other words, the rules. The most important of these rules describes the goal of the game: to place the basketball in the basket, thus scoring a goal. Rules constrain how players might accomplish this: A player can walk as long as he or she dribbles (bounces) the basketball with one hand, can take 3 steps upon placing both hands on the ball, and must pass to a teammate or shoot the ball once he or she is unable to dribble the ball. Other details include the meaning of the lines on the court, definition of acceptable and unacceptable play as well as consequences of unacceptable play. 

The economy can be described much like a basketball game. What are the objects of the game? We have already learned that these include economic goods, entrepreneurs who compete to provide these goods, and consumers who compete to attain these goods. The play of each actor is constrained by property rights.  Ownership of a good allows one to control its use. The goal of the game is to create value. In the world of monetary exchange, this amounts to earning a monetary profit for entrepreneurs. For consumers this manifests as choice aimed at maximizing value derived from one's budget constraint. You have already learned much about the game of exchange and production in the previous chapters.

From both games, concepts emerge from play. In basketball, strategies are used by players on a team. These include special moves, such as a "fake" where a dribbler uses deceptive body language to mislead a defender or an "alley-oop" where one player pass the ball near to the hoop where a receiving player catches it in mid-air and places it in the basket. These concepts are enabled by the rules of basketball. Likewise, entrepreneurs use their creativity to discover new configurations that provide new goods to market or lower the cost of production. Entrepreneurs can profit from these innovations thanks to the rules of the game of exchange.

Statistics emerge from the rules of play. These can be used to evaluate player performance. They can also be aggregated across players to reveal higher level phenomenon. In basketball, this includes the percentage of shots that are successful, the number of fouls incurred by a player, and so forth. These might be aggregated by team to evaluate the team's success and effects of particular strategies. In the game of exchange, these include profits of firms or revenue generated in a particular sector or across the entire economy. Macroeconomics attempts to reveal features of the economy at high levels of abstraction. How much money circulates in the economy? What is the value of production? By how much did prices change? ## Two Ways of Conceptualizing Economic Activity

As you may have observed in the discussion of microeconomics, economists tend to move between different frames for analyzing economic activity. In one case, economic concepts like opportunity cost, diminishing marginal utility, and relative prices provide a theory that can support a narrative approach. While the delineation of a rigorous intellectual pedigree is beyond the scope of this book, we recognize some major contributors to this perspective and the  more technical, aggregate perspective presented here. We may recognize, as Adam Smith did, that the move toward specialization tended to increase efficiency, thereby promoting the creation of more wealth in aggregate. 

We will also discuss the creation and growth of value in the economy using a carefully crafted narrative. Many economists, including Nobel Prize winners Douglass North and Friedrich Hayek, follow this approach. This approach was utilized by early economists who preceded technical developments that enabled more rigorous quantitative analysis and has continue to be refined since the Carl Menger's first presented his Principles of Economics in 1871. Ultimately, all economic analysis, to the extent that it is useful, must be capable of being integrated into a theoretically grounded narrative. More recently, development economists like Hernando De Soto, William Easterly, and Daron Acemoglu and James Robinson have contributed in this manner.

The other approach to macroeconomics is mathematically intensive. Many of the same economists who have participated in the first paradigm also developed work that uses this paradigm. Others, including Robert Lucas, Paul Romer, and Robert Solow have made extensive contributions to this paradigm. Much of the foundation of this sort of work was developed by and in debates that included Milton Friedman and many students from the Chicago school.

The remainder of the book will attempt to weave these two approaches together to create a coherent view of macroeconomic theory.

## The Equation of Exchange

In learning microeconomics, the tendency is to start from a small scale - the individual - and build up to larger concepts and applications. In macroeconomics, we start by defining the entire economy according to divisions that emerge in the economy. Traditionally, this leads to the division of the economy into different sectors: money, capital, and labor markets as well as an aggregate representation of goods and their prices and a market for lending. In this text, we arrange these sectors hierarchically, using the equation of exchange as an organizing heuristic.

The supply side of the macroeconomy represents real goods and services. They represent output when they are created. These goods and services represent value either directly or as generated in the course of exchange. To the extent the goods provide value to consumers, the output translates to income. Thus, nominal income (MVMV) and the total value of output (PyPy) are identical. For the remainder of discussion we will refer to both Py as nominal income and MV as aggregate demand. The value of nominal income is observed directly as the dollar amount of goods and services sold. Using the equation of exchange, we represent nominal income, often referred to as nominal gross domestic product. A common estimate of nominal income is gross domestic product (GDP).  A measure of nominal income is represented as the product of the price level (PP) and real income (yy) by the equation of exchange.

While there are many ways to calculate GDP, we will concentrate on the expenditures approach. We can think of GDP as representing the total money spent on final goods and services.  This approach convenient because it is intuitive.

The expenditure approach defines GDP:

> $GDP = C + I + G +X - M$
>
> $C = Consumption$
>
> $I = Investment$
>
> $G = Government Spending$


GDP within a nation consists of the value of consumption, investment, government spending, and net exports ($X-M$). Consumption represents the aggregate value of goods purchased for final consumption. Investment is the value of savings that is converted into investment in capital. Investors borrow money from savers and attempt to generate profit by their investment. Government spending include the value of taxes and government borrowing. These expenditures can conceivably be described in terms of consumption and investment. However, private individuals do not directly control how government revenues are spent. Exports are goods that are produced within a country but sold abroad. Imports are goods that are produced abroad but sold within the country of interest. 

In this description, GDP is the net value created within a geographic region. Imports are deducted from measures of GDP, even if they are ultimately consumed. The meaning of this formulation can be confusing for those in the process of learning macroeconomic theory. We may simplify the construct if we assume a closed economy, meaning that all trading possibilities exist completely within the region under consideration:

$ GDP_{Closed Economy}=C + I + G$

In reality, the only closed economy that we observe is at the global level. It is convenient to think of GDP in in this manner so long as we do not need to consider international flows of goods. As we develop our understanding of the macroeconomy, we will assume that the economy being analyzed is a closed economy.

## Real Income

GDP measures the value of good sold according to their price. While this can be useful if prices on average remain stable, prices often move together as a result of changes in the quantity of money or changes in the overall level of real income. Movements in the average level of prices, the price level,  shift the level of nominal income.

> Figure 2: Real Gross Domestic Product

One way to separate changes in real income from changes in the price level is to simply count changes the total number of real goods. This would be easy if the economy had only one good. You could add up the quantity produced in order to track changes in real income. In an economy with multiple goods, the value attributed to each good must be estimated and the total summed. One strategy to track real income is to choose a particular basket of goods and track changes in the amount of the goods produced while fixing the value attributed to each unit. This is known as fixed-weighting. In order to estimate changes in real income, prices from the base year are used to calculate the value of goods purchased as income in later years. 

Suppose that we want to estimate the real value of income in a given year. This measurement must be relative to other years. We choose a base year to freeze the value attributed to goods produced. If this year was 2015, then nominal in 2015 would be the same as real income. We generalize this:

$$RGDP_{year} = \sum_{j=1}^{n}p_{j,baseyear}q_{j,year}$$

Real GDP is calculated according to prices of goods in the base year. The product of the quantity of a particular good sold and the price of that good in the base year represent the value contributed by that good to real GDP. 

Suppose that there are only two goods in the economy: pizzas and cars. Suppose that prices and quantities produced of pizzas from 2015 to 2017 were as follows.

>$p_{p,2015} = 10.00$   
>
>$p_{p,2016} = 11.00$
>
>$p_{p,2017} = 12.00$
>
>$q_{p,2015}=1000$
>
>$q_{p,2016}=1050$
>
>$q_{p,2017}=1100$

And the prices and quantities produced of cars from 2015 to 2017 were as follows.

>$p_{c,2015} = 10000.00$   
>
>$p_{c,2016} = 12000.00$
>
>$p_{c,2017} = 12500.00$
>
>$q_{c,2015}=50$
>
>$q_{c,2016}=60$
>
>$q_{c,2017}=70$

If we choose a base year of 2015, then we can calculate $RGDP$ for year i:

$$ RGDP_i=(p_{p,2015})(q_{p,i}) + (p_{c,2015})(q_{c,i})$$

For each year, we calculate $RGDP$:

>$RGDP_{2015}=(10.00)(1000) + (10000.00)(50)=510000.00$
>
>$RGDP_{2016}=(10.00)(1050)+(10000.00)(60)= 610500.00$
>
>$RGDP_{2017}=(10.00)(1100)+(10000.00)(70)=711000.00$

These estimates would accurately reflect changes in value creation if several conditions hold: 

>1) consumer valuations of these goods relative to other known goods do not change 
>
>2) demand for all goods in the basket changed in proportion to changes in income
>
>3) no new category of goods are produced and subsequently demanded by consumers.

The conditions tend not to hold perfectly, but consumer preferences do tend to be stable.  

Another method accounts for changing consumer preferences estimates real income while adjusting for changes in weighting and value over time. Real income ultimately attempts to measure the value of goods produced. Even if the price level remains constant, changing valuation of goods presents a problem for measuring value in terms of a basket of goods that is created. One way to confront this problem is to adjust the base prices of goods in the basket in order to reflect these changes in valuation by consumers. Such estimates of real income are chain-weighted.

Whether real income is estimated using fixed weights or adjusted weights, our concern is that we may refer to the total real value generated in an economy. Economic growth entails the increase of this real value. 

## Price Level

The calculation of $RGDP$ separates prices from the goods and services produced and purchased in the economy. Thus the calculation of $RGDP$ implies a price level, which may be thought of as the average weighted price among a basket of goods.

The price level implied by the calculation of $RGDP$ is known as the $GDP$ deflator. We can use data from our economy of pizza and cars to calculate the $GDP$ deflator. Nominal $GDP$ is defined as $Py$:

$$GDP=Py$$

For our previous example, Nominal $GDP, or simply $GDP$, values are thus calculated:


>$GDP_{2015}=(10.00)(1000)+(10000.00)(50)=510000.00$
>
>$GDP_{2016}=(11.00)(1050)+(12000.00)(60)=731550.00$
>
>$GDP_{2017}=(12.00)(1100)+(12500.00)(70)=888200.00$

Since we know the values of nominal GDP and real GDP, we can calculate the implied price level. Using the values that we have calculated yields:

>$510000.00=P_{2015}(510000.00)$
>
>$731550.00=P_{2016}(610500.00)$
>
>$888200.00=P_{2017}(711000.00)$

Isolate the price level for year *i* to calculate the GDP deflator for that year:

>$P_{2015}=\frac{510000.00}{510000.00}=1.00$
>
>$P_{2016}=\frac{610500.00}{731550.00}=1.198$
>
>$P_{2017}=\frac{888200.00}{711000.00}=1.249$

Although real GDP increased each year, rising prices of the goods purchased also contributed to increases in nominal GDP.
We may construct a price level by fixing the weight attributed to the price of each good. This may be necessary to account for factors that dramatically change the quantity of a good produced and consumed, particularly when such a factor is temporary. The consumer price index  (CPI) is calculated in this manner.
To calculate CPI, we divide the sum of expenditures implied by current prices and the quantities sold of each good in the base year by total expenditures for the basket under consideration.

$$ CPI_{i}=\sum_{j=1}^{n}\frac{p_{j,i}q_{j, i_0}}{p_{j,i_0}q_{j,i_0}}$$
>$i=year$ $i$
>
>$i_0=base$ $year$
>
>$j=good$ $j$

Using this formula, we derive CPI for our pizz and car economy:

>$P_{2015}=\frac{(10.00)(1000)}{(10000.00)(50)}=1.00$
>
>$P_{2016}=\frac{(11.00)(1000)}{(12000.00)(50)}=1.198$
>
>$P_{2017}=\frac{(10.00)(1000)}{(10000.00)(50)}=1.249$

Although these measures are very close, it is possible that the measures yield different outcomes. The CPI will tend to overstate inflation, whether positive or negative, when magnitudes of changes in prices are especially large. The CPI fails to account for substitution away from more expensive goods toward similar goods that are relatively cheap. Suppose that peanuts are included in the CPI basket and that peanut yields suffered from pest infestation. Consumers might substitute toward other butters such as cashew butter or almond butter, and therefore not directly suffer the consequences of rising prices. The CPI, however, would record this inflation using the weights derived from the base year. The GDP deflator, on the other hand, would not account for the rise in price of peanut butter to the extent that individuals substitute from the good. Thus, the CPI tends to overstate inflation and the GDP deflator tends to understate inflation. While GDP deflator is preferable from a theoretical perspective as it emerges directly from present consumer preferences, the usefulness of either index depends on the ends of the investigation.

## Aggregate Demand

Similar to individual markets, a high level view of the macroeconomy identifies demand for and supply of all economic output. The equation of exchange represents these as aggregate demand ($MV$) and aggregate supply ($y$). Aggregate supply represents real value created in the economy. It is priced in light of the average level of prices, $P$.

The demand side of the economy is represented as expenditures denominated in currency. We assume that the economy analyzed is a monetary economy where all goods are sold for money and, therefore, not bartered. In order to be sold, a consumers must offer the currency equal to the value of the sale price of goods. Money is on one side of every exchange that involves real goods and services ($y$). As you might imagine, the quantity of money available influences the level of average prices.

Just as aggregate supply is comprised of two variable, $P$ and $y$, aggregate demand is comprised of $M$ and $V$. Total expenditures is equal to the quantity of money multiplied by the average number of times each monetary unit is spent over a given time period. In later chapters we will show that there are different types of money. For now, we will concentrate on base money. 

## Supply of Money

The variable $M$ represents the quantity of money in circulation. If there exist 1 trillion dollars in circulation, then the value of $M$ is 1 trillion dollars. Later we will consider different types of money. For now, we will consider the role of base money, sometimes referred to as outside money.

Base money is what you likely think of when you imagine money. It includes paper currency printed by the central bank. It also includes credits from the central bank. This modern manifestation is not easy to grasp by intuition. We like to imagine that money is ultimately physical and that claims to money are ultimately claims to a physical good known as money. This was true during the operation of commodity standards, many of which used gold, silver, and copper. Over the course of the last two centuries, governments have transformed this conception of money, exercising increasing control over the monetary system. They began by limiting the ability of citizens to use the metal as money, instead issuing notes that were supposed to backed by rare metals. Over time, these restrictions tended to grow until the international gold standard was fully ended in 1971 by Richard Nixon.

There are two way to conceptualize the supply of money. One is to treat money as a static quantity that is adjusted exogenously. In modern context, the quantity of money in circulation is determined by the decisions of the central bank. To expand the quantity of money, the central bank creates new money and either spends or loans this money. To shrink the money supply, it sells assets or allows these loans to expire without displacing these losses with newly created money. 

The quantity of money may also respond to demand for it. In the case that it does, it is endogenously governed by market processes. We see this in the modern environment with expansions and contraction of credit. During a boom period, returns to investment tend to be relatively high. Banks create more loans as expectation of higher returns grows, thus increasing the total supply of money supported by the base money stock. Under a commodity standard, increases in demand for money increase its value. This makes mining of precious metals more profitable leading to an increase in the quantity supplied to the market.

## Demand for Money

Regardless of the mechanism that governs the quantity of money, the influence of money on prices and income depend upon demand for money. There are two types of demand for money: portfolio demand for money and transactions demand for money. Portfolio demand represents demand to hold money over the duration of the period observed. Likewise, transaction demand for money represents demand to spend money during the same period.

Total money demand is the combination of transactions demand and portfolio demand. Since $Py$ is transactions demand, then ($\frac{1}{V}$) must be portfolio demand. The average proportion of the money stock held is the inverse of its average rate of expenditure. We can rewrite the equation of exchange to represent these to forms of demand:

>$M_d = \frac{Py}{V}$
>
>$Py=Transactions$ $Demand$
>
>$\frac{1}{V}=Portfolio$ $Demand$

We may rewrite the inverse of velocity as k, portfolio demand:

>k=\frac{1}{V}

Thus yielding,

>$M_d=Pyk$

Since $M_s=M_d$ in equilibrium, we define the money stock:

>$M_s=Pyk$

The product of the quantity of money and the velocity of money represents total expenditures in an economy. The value of total expenditures is always equal to nominal income. Thus, if the quantity of money is $1 million and the velocity of money is 3, then nominal income must be $3 million. Total expenditures matches the value of nominal income.

## Price of Mnoey

As with any other good, the price of money is determined by demand and supply for it. The difficulty with defining the price of money is that there are at least as many prices of money as there are markets where money is exchanged. If a pizza sells for 10 dollars, than 1 dollar is worth 1/10 of a pizza. Measuring the price of money in this manner can be useful. The increase of a price of any good represents a decrease in the price of the dollar and vice versa. 

It is easier to think of the price of money in terms of its average price. Just as the price of money is the inverse of the price of the good exchanged for money, the average price of money is the inverse of the price level.

>$P_M=\frac{1}{P}$
>
>$P_M=Average$ $Price$ $of$ $Money$

The price of money might therefore be measured by changes in the price level. 

> Figure 4 GDP Deflator and Trade Weighted U.S. Dollar Index

Changes in the price of money might also be detected by changes in the price of a given money relative to other currencies. For example, suppose that the the U.S. dollar devalues against the British pound. The price of the pound measured in dollars increases. Likewise, the price of the dollar in terms of pounds decreases. Exchange rates between two currencies can be volatile. The Trade Weighted U.S. Dollar Index provides a relatively stable measure of the dollar in terms of a basket of other major currencies. 

A measure of one currency against another tends to be more volatile as currency can be traded at relatively lower cost than goods. Thus exchange rates tend to overshoot long-run equilibrium values compared to estimates of the price level which exhibit greater stability. Exchange rates can also fail to detect inflation if all currencies being considered experience concurrent devaluation. The general trend of a weaker trade weighted dollar does match the trend of an increasing price level.

## Logs and Interpretation of Macroeconomic Data

The graphs of GDP and real GDP, and many other estimates of macroeconomic data, look like hockey sticks. It is not appropriate to present macroeconomic data without adjusting the data to reflect rates of change. For example, an increase of real GDP by $2 billion is a dramatic increase if real GDP is $10 billion. However, if real GDP is $1 Trillion, a $2 billion dollar increase is quite small. We can account for the relative magnitude of these changes by comparing logged values of macroeconomic variables.

## Dynamic Equation of Exchange

It is useful to relate changes in the equation of exchange in terms of rates (percentage) rather than raw values. It is possible to convert the values of the equation of exchange into rates using a logged derivative.

First, a brief review of logarithms are in order. We will be using natural:

$$ln(x)=log_e(x)$$

To take the log of $x$ is to ask to what power the base must be raised to yield the value of $x$.

$$e^\theta = x$$

Using this notation, we identify that:

$$ln(x)=\theta$$

Conveniently, we can ask what is the natural log of 1 and what is the log of e. Since any number that is not zero that is raised to the 0 power is 1:

$$ln(1)=0$$

A number raised to the first power yields the same number:

$$ln(1)=e$$

Logged values have the same algebraic properties as exponents with identical bases. The product of identical bases with exponents yields the base raised to the sum of these exponents:

$$x^\alpha x^\beta = x^{\alpha + \beta}$$

Similarly, dividing bases with exponents yields the base raised to the difference of these exponents:

$$\frac{x^\alpha}{x^\beta}=x^{\alpha - \beta}$$

This may also be represented by use of negative exponents:

$$x^\alpha x^{-\beta}=x^{\alpha - \beta}$$

Following similar logic, a log of a product is the same as the sum of each log:

$$ln(xy)=ln(x)+ln(y)$$

Likewise:

$$ln(\frac{x}{y}) = ln(x) - ln(y)$$

Finally, the derivative of a log yields the rate of change:

$$\frac{d(ln(x))}{dx}=\frac{1}{x}dx = \frac{dx}{x}$$

$$MV\equiv Py$$

$$ln(MV)=ln(Py)$$

$$ln(M)+ln(V) = ln(P)+ln(y)$$

Next we take the total derivative of the logged version of the equation of exchange. Since these are sums, the transformation is intuitive:

$$\frac{dln(M)}{dM}+\frac{dln(V)}{dV}\approx \frac{dln(P)}{dP}+\frac{dln(y)}{dy}$$

$$\frac{dM}{M}+\frac{dV}{V}\approx \frac{dP}{P}+\frac{dy}{y}$$

Taking the derivative of a log estimates the change in the variable divided by the variable itself or, in other words, the rate of change. We have already estimate one of these rates. Inflation is the rate of change of the price level. As we develop our macroeconomic framework, the use of rates will create consistency in framing. It also will provide better understanding macroeconomic statistics that are presented as rates.
