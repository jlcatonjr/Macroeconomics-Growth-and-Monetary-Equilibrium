# Data Homework Template

### This is a guide to help you complete the data homework assignments.

We will import functions from the datlib folder. these will be used to import data and plot the data. You only need to understand how to enter input for these functions.

To import data, we will need to specify the codes for the data and the range of dates for which we will collect data. 

```python
#plots.py
# if you have not installed pandas-datareader, uncomment the command 
# in the next line. After intallation, you can remove the command
#!pip install pandas-datareader

import datetime
from datlib.FRED import *
#Import monetary base
data_codes  = {"Base: Total": "BOGMBASEW",
               "Base: Currency in Circulation": "MBCURRCIRW",
               "Excess Reserves": "EXCSRESNW"}
# Select start and end dates
start = datetime.datetime(1975, 1, 1)
end = datetime.datetime.today()
```

```python
# freq refers to data frequency. Choose "D", "W", "M", "Q", "A"
#                                       day, week, month, quarter, annual
# a number may also be place in front of a letter. "2D" indicates
#       alternating days
data = gather_data(data_codes = data_codes, start = start, end = end, freq = "M")
```

You can view the data by calling the entire dataframe.

```python
data
```

Or you can call specific rows by the order in which they appear. For example, if I want to call rows 0 through 9, call _data.iloc[start:last + 1].

```python
data.iloc[0:10]
```

If you want to call dates from a certain year, call data.loc[year]. Make sure that the year is in quotations.

```python
data.loc["1975"]
```

You can call a range of years using _data.loc[start:finish]_.

```python
data.loc["2007":"2008"]
```

# Selecting columns

Instead of calling an entire dataframe, you can call a single column by passing a string (i,e., use quotation) the column name surrounded by brackets.

```python
data["Base: Total"]
```

You can also call multiple columns by passing a list of column names.

```python
data[["Base: Total", "Base: Currency in Circulation"]]
```

Finally, you can define the list before passing it. This will make it easier to call the same column names again later.

```python
names = ["Base: Total", "Base: Currency in Circulation"]
data[names]
```

# Plotting data

### Line Plots

Now that the data has been imported, we can visualize it using plot_lines(), which we import from datlib.plots

```python
from datlib.plots import * 
# set default fontsize
plt.rcParams.update({"font.size":32})
plot_lines(df = data, linewidth = 1, figsize = (40, 20))
```

You might only want to plot one column of data. In that case, just pass the dataframe that calls that one column.

```python
data_to_plot = data["Base: Total"]
plot_lines(df = data_to_plot, linewidth = 1, figsize = (40, 20))
```

The plot_lines() function also includes an option for a secondary y-axis.

```python
plot_lines(df = data, linewidth = 1, figsize = (40, 20), secondary_y = "Base: Total")
```

### Stacked Plots

Suppose that you want to see how the values of different components of an aggregate change over time. For example, the total base is composed of currency in circulation and reserves. Let's see how much of the base is comprised of these values. The remaining value, required reserves, is implied by the gap between the dotted line and the stacked plots.

Unlike in our line function, the *plot_stacked_vars()* function does not simply plot every value in the dataframe. Instead, we must define which values will be stacked and which variable represents the value of the entire aggregate.

```python
stack_vars = ["Base: Currency in Circulation",
             "Excess Reserves"]
aggregate_var = "Base: Total"
plot_stacked_lines(data, stack_vars, linewidth = 1, 
                       figsize = (40, 20),
                    total_var = aggregate_var,
                      title = "Components of the Monetary Base")
```

### Scatter Plots

You may also create scatter plots by selecting two or more variables. The _plot_scatter()_ tracks time by using color.

```python
plot_scatter(data=data, s = 75, figsize = (40, 20), save_fig = False, pp = None)
```

### Creating New Columns

Suppose that you want to create a new variable: the fraction of the monetary base that is currency in circulation. To do so with _pandas_ is quite simple. You can accomplish this in one of two ways. 

You can use the _.div()_ method from pandas: data[var1].div(data[var2])

```python
data["Excess Reserves / Base"] = data["Excess Reserves"].div(data["Base: Total"])
```

Alternately, you can just use the math operators instead of *.div()* or any of the other dataframe methods (i.e., *.add()*, *.sub()*, *.mul()*, *.pow()*)

```python
data["Excess Reserves / Base"] = data["Excess Reserves"] / data["Base: Total"]
data["Excess Reserves / Base"]
```

```python
plot_lines(df = data, linewidth = 1, figsize = (40, 20), secondary_y = "Excess Reserves / Base")
```
