# Purposive Action
## Unraveling the Mystery

> <center> <b> Figure 1
    
There is order to the world. Step on to the streets of New York City and you will notice that traffic of vehicles and persons follow identifiable patterns. Pedestrians wait for the last car in a line of traffic to pass before they start crossing. The lights in Times Square flash advertisements to potential customers. Step out of a farmhouse in rural Montana and you will notice a different sort of order. Crops are neatly planted. Storehouses await the coming yields of harvest time.[1]

What is common between these two settings? Clearly there is order, but what is the source of this order? This book argues that the source of order is the human drive toward value creation. Every action embodies an attempt to create value in the eyes of the actor. If you lived in a world where you were the only person, then your every act would directly yield value for yourself. In a world with many people, there exist opportunities to improve one's state by creating value for and with one another. A society becomes prosperous when individuals are allowed to interact in a manner that benefits one another. This may occur through serial reciprocation, as in a friendship where two people share food, exchange gifts, and engage in thoughtful conversation with one another. Or, in the case of a market setting, this may occur through the creation of products and provision of services intended for use by complete strangers. These are sold for money that serves as a means to attaining the ends that one ultimately seeks.

The overarching argument of this book can be summed as follows: 

- Human action is purposive. 
- The purpose of action is to create value that reflects the preferences of the person acting.
- Value is created when an individual realizes a state of the world that he or she prefers more to the one that existed or that the individual expects would otherwise exist.
- Because scarcity is an inherent feature of the world, men and women structure their actions, affiliations, and ownership of assets in a manner that they expect will create the most value in accordance with their preferences.

Many describe economics as the study of the allocation of resources among competing uses. Competitive markets promote the allocation of resources that generates the greatest value measured in terms of dollars. Such resource allocation is certainly included in the study of economics, however, it is a consequence of the purposive action of a plurality of actors. Economics is the study of value and action intended to realize value. In this chapter we will briefly introduce the concept of value and a system of understanding naturally evolves from a focus on value driven action.

Our study of the macroeconomy will integrate the same suppositions as above. We may add some other fundamental issues that drive analysis of the macroeconomy:
- Economic growth depends fundamentally on the development of human knowledge
- Changes in the supply and demand for money impact activity throughout the economy 

## From Whence Value?

The economic problem centers around value. It is no coincidence that the most significant turning point in economics was the result of development in the theory of value. 

At the end of the 18th century, economists were still grappling with the problem of value. Adam Smith identified this problem, known as the "diamond-water" paradox. The question was simple, if value is derived from a good's usefulness, then why are diamonds relatively expensive compared to water? 

> <img src="https://github.com/jlcatonjr/Macroeconomics-Growth-and-Monetary-Equilibrium/blob/main/Chapter%201/Figure%202.jpg?raw=true" alt="City" />
> <center> <b> Figure 2
> 
> Nothing is more useful than water: but it will purchase scarce any thing; scarce any thing can be had in exchange for it. A diamond, on the contrary, has scarce any value in use; but a very great quantity of other goods may frequently be had in exchange for it. (Smith 1776, 1.4.13)
    
Without water, no person could last more than a few days. "Labour", Smith wrote, "is the real measure of the exchangeable value of all commodities. The real price of everything, what everything really costs to the man who wants to acquire it, is the toil and trouble of acquiring it (1.5.1-2)." He noticed that the labor required to produce goods was correlated with the price of the good produced. Smith observed:
    
> If among a nation of hunters, for example, it usually costs twice the labor to kill a beaver which it does to kill a deer, one beaver should naturally exchange for or be worth two deer. It is natural that what is usually the produce of two days' or two hours' labour should be worth double of what is usually the produce of one day's or one hour's labour. (1.6.01)

Ricardo extended Smith's labour formulation from a theory of value for commodities and for a primitive society to a more general theory of value. While Ricardo recognized that the labour theory of value was an imperfect one,  he "shrugged his shoulders" at the problem (Blaug 1992, 96).  Without an appropriate theory of value, the classical economists were unable to explain economic coordination in a systematic manner.

Smith and Ricardo were observing an outcome of the market process. Competition between firms tends to minimize the cost of production as high-cost producers are removed from the marketplace. Under these conditions, the cost of providing a good tends to equal the price charged for the good. It is unable to describe changes in valuations. If suddenly a dear skin becomes 4 times as valuable as before due to a change in preferences and scarcity of deer, the price of the deer skin in terms of beaver skins will deviate from the ratio of labor costs reflected in the old price that Smith describes. Classical economists were not able to describe short-run fluctuations that escaped the logic of the labor-theory of value. Thus arose the association of the classical school of economics with long-run analysis (Menger 2007, 8).
    
## Marginal Analysis
    
A rigorously worked out improvement on the labor theory of value would have to wait until 1871, the year of the marginal revolution in economics. That year, three economists - Carl Menger, William Stanley Jevons, and Alfred Marshall - separately formulated economic systems that employed marginal analysis. Value, according to these systems, was dependent upon the utility provided by possession of an additional unit of an object desired by the acting agent.
In the early pages of his foundational work, Principles of Economics, Carl Menger places subjectively determined value at the center of his framework. For Menger, value is not given in terms of the cost of labor. Nor can it be taken for granted in analysis. It must be defined. 
    
All things are subject to the law of cause and effect. . . It is impossible to conceive of a change of one's person from one state to another in any way other than one subject to the law of causality. (1976, 51)
    
Economics views all action in terms of means and ends in light of the logic embedded in an agent's understanding. It presupposes that the world is ordered, subject to the forces of cause and effect (Mises 1949, 27). Rational beings are capable of acting in a world in a manner that transforms it and one's state in it. In the process, acting men and women come to posses, employ, and consume artifacts that we refer to as economic goods: objects subject to scarcity that are imbued with value by the actions of persons that come to possess them. The economic good is a fundamental category of object in Menger's analysis:
    
> Things that can be place in causal connection with the satisfaction of human needs we term useful things. If, however, we both recognize this causal connection, and have the power actually to direct the useful things to the satisfaction of our needs, we call them goods (1976, 52).
    
The value of a good is evidenced by the opportunity cost associated with its acquisition or realization. This value is derived from individual desire. In economics, we refer to this as an individual's preference. Economic theory holds that agent action can be described by a preference ordering that determines choices that the agent makes as he or she interacts with the environment.
    
## Opportunity Cost
    
An individual chooses one end over another. He must choose a means to attain the desired end. This person chooses one means over another. Action is most often intervention into the existing state of affairs. By action, one attempts to realize a desired state of the world compared to the state that would arise without intervention. Consider that you are currently reading this book. Perhaps if you were not reading this book you would watch an episode from your favorite television series or spend time with your best friend over a cup of coffee. Whatever action you would expect to choose had you not read this book is your opportunity cost.
    
We assume that a person chooses to take the action that he desires more than others available. The value derived from action and realization of one's desired state is expected to exceed the costs incurred for its attainment. Every person attempts to maximize value ex ante. He takes action that, preceding the action, he expects will yield results that he prefers.

Every action must take into account a scarcity of resources that is everywhere present in human society. The notion of opportunity cost assumes that an individual does not have at his disposal unlimited resources. And even without scarcity of resources, there will always exist scarcity of time. All  agents are subject to a budget constraint that limits the scope of choice and means of their influence over the state of affairs in the world.
    
Means, ends, and opportunity cost implied in economic analysis drive many of the macroeconomic patterns that likely interest you. Why do prices tend to fall on average during an economic depression? Why do firms invest in research and development rather than producing more of the same goods? Why do wages tend to be high in developed countries like the United States, Great Britain, and Australia? Economic analysis provides the key to answering these questions and others. It is a window into social orders.


## Comparative Advantage

Opportunity cost enables us to consider comparative advantage. Ricardo first spoke of comparative advantage in his discussion of wine and cloth production in Spain and England respectively (1817). He argued that nations were better off if individuals were allowed to produce the goods that they can produce most cheaply compared to producers in other nations, and to trade for those goods that others produce more efficiently. If wine can be produced profitably in Spain and cloth in England, then English cloth manufacturers should purchase Spanish wine and Spanish wine producers should purchase English cloth. 

Ricardo's logic applies to any trade, not just cloth and wine, but all types of productive activity. Consider the Stephen Curry who plays for the Golden State Warriors. Stephen Curry is a talented leader and basketball player. He seems to make best use of this talent playing for the Golden State Warriors. There, his presence is a unifying force that has helped move the team to win a championship and a heartbreaking miss last season. The opportunity cost of Curry’s participation with the Golden State Warriors is likely participation with another team. The value he creates for the Golden State Warriors is substantial enough that the team has retained him since his entrance in 2009. Although Curry would likely be valuable for any team on which he plays, he has a comparative advantage playing for the Golden State Warriors. This becomes more obvious if Curry were to change his occupation altogether. Curry may be a fantastic manager. When he graduated from college, he could have entered a field that allowed him to exploit his leadership ability through management. Even if Curry was one of the best managers, it would be difficult for him to surpass his current pay and other benefits he receives playing for the Warriors. In 2017 alone, he earned $34.68 million! This is to say that Curry may have an absolute advantage in comparison to others both in his management ability and his ability on the basketball court. Market prices, however, inform Curry that the most valuable use of his skills will leave him playing for the Warriors, even if he had an absolute advantage as a manager and as a basketball player, he has a comparative advantage in playing basketball.

Comparative advantage allows us to explore areas that may, at first glance, seem to be a source of waste or laziness. Bill Gates, who is estimated to be the world’s wealthiest person in 2017, generates a tremendous amount of value with his time. It makes little sense for Bill Gates to spend time preparing meals or cleaning his own home. Every moment he spends engaged in this activity, he could be dedicating to create more value elsewhere. Cleaning and cooking is especially costly for Bill Gates. He is better served by purchasing cleaning and cooking services than by performing these tasks himself. He may choose to do this simply out of enjoyment of the activity or belief in the personal enrichment from such activity. In terms of value as measured by money prices, he is incurring great cost if he chooses to do so.

Development of specialization that arises due to comparative advantage plays a significant role in the provision of knowledge in society. Specialization itself allow individuals to develop knowledge of a particular category of activity. Specializing as an athlete, Curry has been able to develop his skills, teamwork, and understanding of basketball in a way that would not be possible if he could only play basketball in his spare time. The same logic holds for the development of applied knowledge that has radically transformed the world over the last half century. High speed computer processing and automated transportation are possible because individuals who produce them are able to dedicate their productive energies and learning to a small number of tasks. The pace of change appears only to increase as freedom to develop technology and organization also increases.

## Specialization
How does specialization also enable the development of technology and capital? Economists often use the story of Robinson Crusoe to convey the improved efficiency entailed in specialization and exchange. Imagine two castaways stranded on an island.  Crusoe arrive first. Stranded on an island, he must find the means to satisfy his desire for food and shelter. He cannot work simultaneously to achieve both. Thus, he must spend some portion of his time fishing or harvesting coconuts and another portion to dedicated to building and maintain shelter. As long as Robinson Crusoe is alone, it is difficult for him to develop means for either. When Friday arrives on the island, he brings with him similar demand for food and shelter.

The economic problem becomes more complex and, therefore, provides more possibilities. If Friday enjoys fishing, he can spend his time gathering food in this manner while Robinson Crusoe concentrates his energies on providing shelter. Perhaps Friday can catch enough fish for Robinson Crusoe in 6 hours. If he likes, he can spend the rest of the day resting. However, since Robinson Crusoe is working on the other side of the island, he may feel motivated to produce more fish to exchange, leaving both himself and Robinson Crusoe better off. Suppose that Friday would like to be able to catch more fish in less time, so he spends his free time building a net. At first, this is a laborious process as he lacks the knowledge for this, so he must spend time experimenting. Perhaps Friday needs two months two develop a proper net. Once he has discovered a successful strategy for building nets, he can either catch more fish or dedicate less time to fishing. Further, the time taken to create a new net will be less as it will not require the initial costs of discovery that Crusoe first faced.

The welfare of both parties will continue to improve so long as technological advancement is possible. One change enables another. The development of a more cost efficient form of production can support larger populations and more complex societies. Before the development of agriculture, it may be difficult for metallurgists to survive. As society grows more wealthy, both the variety and level of consumption increases. To the extent that one can develop and maintain a social position for oneself with a specialization such as this, he can maintain his well-being through production and exchange. Likewise, this change in society opens the opportunity for carpenters who have specialized knowledge of constructing and repairing buildings. Underlying this change is the fundamental principle we identified early on: humans act to improve the state of the world as they perceive it.Decision to action occurs in light of the costs and benefits that such action is expected to generate. Humans rarely imagine radical change. Even that thought be radical is usually marginal in light of the present state. Thomas Edison’s discovery of the light bulb, while unexpected by many, required a combination of elements that were readily available to him. This innovation required a search through the vast space of possibility that exists at the fringes of existence. Once found it may enable and encourage many other innovations that were not previously possible. The development of the light bulb, for example, made valuable the provision of electricity in cities. This further provision opened up many opportunities as it allowed for technical development and specialization that would have otherwise been impossible.

## Technology
So far, we have discussed the development of the exchange game. Though it was not described in such terms, the development of this game represents the mutual creation of social technology by the two castaways in our story, Robinson Crusoe and Friday. Each must learn to bargain explicitly and implicitly with one another. From this bargaining emerges prices that reflect preferences of each trader in light of the costs of supply of their desired goods. Prices are artifacts that help both parties to form expectations concerning the value that their own goods will fetch in the market place. This bargaining and the prices that emerge from this process increase the efficiency of production and exchange, thereby producing value.

More generally, we can think of technology as organization of social and material resources that influence the value generated from an activity. Often, we think of technology in terms of physical technologies. Consider the development and implementation of rail transportation during the 19th century that substantially reduced transportation costs [1]. Railroads were much safer than transit by water. They also operate during winter time, allowing suppliers to economize on storage costs. Further developments in shipping technologies include the creation and distribution of standardized shipping containers in the 20th century [2].

The adoption of one cost reducing technology begets the creation and adoption of another. Railroads enabled the adoption of standardized time zones, an element critical to coordination of train routes and time sensitive projects that cover large regions [3]. In both Great Britain and the United States, the new standard for keeping time was adopted long before central governments legally recognized these innovations. In 1883, American railroad companies adopted standardized time and the system of time zones recommended by William F. Allen. Within a year, most large cities had adopted this standard. Only in 1918 was this system finally recognized by the central government with the Standard Time Act.

Innovations realized during the industrial revolution greatly improved productive efficiency, but such improvements would not be possible without the development of more efficient social technology that brings investors confidence that their large fortunes will not be repossessed by any individual or the state responsible for enforcing the law. That is, the development the industrial revolution required the transformation of government that inhibited exchange through restrictions on trade like high tariffs and import quotas and a general lack of permission for individuals to engage in market activity.

The development of superior governance technology occurred in Britiain in the late 17th century. With the Glorious Revolution 1688, William and Mary established a government that empowered parliament and which tended to respect the rights of merchants more than the previous regime [4]. The government became less dependent upon a person and, therefore, more perfectly the rule of law. The logic governance under the new system was better defined and more predictable. Together, these changes incentivized businessmen to expand commerce. Only by this expansion of commerce was the industrial revolution able to attract the funds necessary to takeoff in Great Britain and, later, the United States. Improvements in the predictability of human interaction through securement of the rule of law enabled the adoption, spread, and development of new inventions – physical technologies.

The maintenance and adaptation of technology, physical and social, are necessary conditions for continued economic development. Social technology that promotes economic growth enables a broad scope of action for all individuals. Without this freedom, we cannot maintain economic growth. In an age where political tumult within and between governments shows potential to degrade the rule of law, we must be vigilant to maintain the integrity of our institutions of governance and their limited, predictable scope of action [5].
