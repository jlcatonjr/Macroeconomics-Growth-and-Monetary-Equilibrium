## Useful Commands: Logs, Differences, and Percent Change

Sometimes, you might have to transform data. Below are a list of common transformations.

### Logging Data

Logging data is useful if you would like to observe change in terms of proportion instead of in absolute units. For example, if you plotted the doubling of a value, starting with $1$ ($2^0$), over $100$ periods, changes in later periods would be significantly larger than changes in earlier periods. But if you log these values, changes will appear linear.

```python
doubling_list = [2 ** i for i in range(30)]
doubling_list
```

Compare this to the change in logged values:

```python
# when you log data, we import numpy and call log command
import numpy as np
logged_doubling_list = np.log(doubling_list)
logged_doubling_list
```

Notice that the logged values appear to increase linearly. We can view these lists side by side to better understand how the logged values represent proportional change. To do this, we will pass a list of lists to a pandas DataFrame constructor.

```python
import pandas as pd
# use df.T to make the list read vertically instead of horizontally.
pd.DataFrame([doubling_list, logged_doubling_list]).T
```

As you may have guessed, we can perform a similar operation by passing a dataframe column to the log function. Let's log the value of the total stock of base money and then call the observed and logged values.

```python
# Log Data
import numpy as np
data["Logged Base: Total"] = np.log(data["Base: Total"])
data[["Base: Total", "Logged Base: Total"]]
```

### Differencing and Percent Change

We may want to explicitly identify rate of change. There are a few ways to do this. The choice of which to use depends on your purpose. Over one or several periods you may

1. Calculate the change in an observed value
2. Calculate the difference of logged values. 
3. Calculate the percent change of a value

Methods 2 and 3 will provide results that are approximately the same.

#### Method 1: Difference Data

To calculate the change in observed values, we can use the a dataframe's *.diff()* method. Let's observe changes in the quantity of base money during each period. The method defualt to checking the change over one period, but you could pass an integer different than 1 to *.diff()* to check changes over longer intervals.

```python
# Difference Data (n periods)
# same as data["Base: Total"].diff(1)
data["Base: Total (Diff)"] = data["Base: Total"].diff()
data["Base: Total"].diff()
```

```python
data.diff()
```

#### Method 2: Difference Logged Data
The *.diff()* method works with multiple columns. Since we also want to see the first difference of the logged values, lets take the difference of both.

```python
data["Logged Base: Total (Diff)"] = data["Logged Base: Total"].diff()
data[["Base: Total (Diff)", "Logged Base: Total (Diff)"]]
```

```python
np.log(data).diff()
```

#### Method 3: Calculate Percent Change
Finally, let's calculate percent change using the *.pct_change()* method. Like with the *.diff()* method, this method considers the percent change over 1 period by default. You can pass a different value for different periods.

```python
# Calculate Percent Change over n periods
data["Base: Total (% change)"] = data["Base: Total"].pct_change()
data[["Base: Total (Diff)", "Logged Base: Total (Diff)", "Base: Total (% change)"]]
```
