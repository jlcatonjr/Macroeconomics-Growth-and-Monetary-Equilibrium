```python
import datetime
from datlib.FRED import *
#Import monetary base
data_codes  = {"Base: Total": "BOGMBASEW",
               "Base: Currency in Circulation": "MBCURRCIRW",
               "Excess Reserves": "EXCSRESNW"}
# Select start and end dates
# datetime.datetime(year, month, day)
start = datetime.datetime(1975, 1, 1)
end = datetime.datetime.today()
```

```python
# freq refers to data frequency. Choose "D", "W", "M", "Q", "A"
#                               day, week, month, quarter, annual
# a number may also be place in front of a letter. "2D" indicates
#       alternating days
data = gather_data(data_codes = data_codes, 
                   start = start, 
                   end = end, 
                   freq = "M")
```

## Next, we are going to learn to plot data

```python
from datlib.plots import *

# here is a simple way to plot data
data.plot()
```

```python
# plot only variables in list: cols
data[cols].plot()
```

```python
# set default fontsize
plt.rcParams.update({"font.size":40})

# call plot_lines function from plots.py
plot_lines(df = data, linewidth = 3, figsize = (40, 20))
```

```python
# call plot_lines function from plots.py
plot_lines(df = data, linewidth = 3, figsize = (40, 20), 
           secondary_y = "Excess Reserves")
```

```python
# call plot_lines function from plots.py
plot_lines(df = data[["Base: Currency in Circulation", "Excess Reserves"]], linewidth = 3, figsize = (40, 20))
```

```python
plot_lines(df = data, 
           linewidth = 3, 
           figsize = (40, 20),
          secondary_y = "Excess Reserves")
```

```python
# chose variables that 
stack_vars = ["Base: Currency in Circulation",
             "Excess Reserves"]
aggregate_var = "Base: Total"
plot_stacked_lines(data, stack_vars, linewidth = 8, 
                       figsize = (40, 20),
                    total_var = aggregate_var,
                      title = "Components of the Monetary Base")
```

```python
plot_scatter(data=data[stack_vars], s = 100, 
             figsize = (40, 20), 
             save_fig = False, pp = None)
```

```python
# data[name] = . . .
data["Excess Reserves / Base"] = data["Excess Reserves"].div(data["Base: Total"])
data["Curr in Circ / Base"] = data["Base: Currency in Circulation"].div(data["Base: Total"])

data
```
