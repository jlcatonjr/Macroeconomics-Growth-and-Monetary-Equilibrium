In this course, we will regularly download data from [FRED](https://fred.stlouisfed.org/). Below is code that we will use to download data from FRED.

```python
import datetime
from datlib.FRED import *
#Import monetary base
data_codes  = {"Base: Total": "BOGMBASEW",
               "Base: Currency in Circulation": "MBCURRCIRW",
              "Excess Reserves" : "EXCSRESNW",
              "Effective Federal Funds Rate" : "EFFR"}

# Select start and end dates
# datetime.datetime(year, month, day)
start = datetime.datetime(1975, 1, 1)
end = datetime.datetime.today()
```

```python
start, end
```

```python
data_codes
```

```python
# freq refers to data frequency. Choose "D", "W", "M", "Q", "A"
#                               day, week, month, quarter, annual
# a number may also be place in front of a letter. "2D" indicates
#       alternating days
data = gather_data(data_codes = data_codes, 
                   start = start, 
                   end = end, 
                   freq = "W")
```

## Next, we are going to learn to plot data

You can plot data by using the canned matplotlib extension for pandas.

```python
from datlib.plots import *

# here is a simple way to plot data
data.plot()
```

If you want to select only a subset of data, you can a list of the specific columns that you wish to plot.

```python
# plot only variables in list: cols
data[cols].plot()
```

I've created a function that you can use to make nicer looking visualizations. Before we use this function, lets increase the default font size

```python
# set default fontsize
plt.rcParams.update({"font.size":40})

# call plot_lines function from plots.py
plot_lines(df = data, linewidth = 3, figsize = (40, 20))
```

We can also plot variables using a secondary axis. Multiple variables can be plotted using secondary axis.

```python
# call plot_lines function from plots.py
plot_lines(df = data, linewidth = 3, figsize = (40, 20), 
           secondary_y = ["Excess Reserves"])
```

We can also create stack plots using a function from the plots.py file in the datlib folder.

```python
# chose variables that 
stack_vars = ["Base: Currency in Circulation",
             "Excess Reserves"]
aggregate_var = "Base: Total"
plot_stacked_lines(data, stack_vars, linewidth = 8, 
                       figsize = (40, 20),
                    total_var = aggregate_var,
                      title = "Components of the Monetary Base")
```

There is also a functionto plot every pairwise combination of variables in scatter plot. The color of the variable indicates period of the observation.

```python
plot_scatter(data=data[stack_vars], 
             s = 100, 
             figsize = (40, 20), 
             save_fig = False, pp = None)
```
