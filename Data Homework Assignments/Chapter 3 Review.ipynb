```python
import datetime
from datlib.FRED import *
#Import monetary base
# MV = Py
# Download: M, Py, y
#  M: Quantity of Base Money
#  Py: Nominal GDP
#  y: Real GDP

# Calculate: V, P
# V: Velocity
# P: Price Level (GDP Deflator)

data_codes  = {"Base: Total": "BOGMBASEW",
               "Base: Currency in Circulation": "MBCURRCIRW",
              "NGDP": "GDP",
              "Real GDP": "GDPC1"}

# Select start and end dates
# datetime.datetime(year, month, day)
start = datetime.datetime(1975, 1, 1)
end = datetime.datetime.today()
```

```python
# freq refers to data frequency. Choose "D", "W", "M", "Q", "A"
#                               day, week, month, quarter, annual
# a number may also be place in front of a letter. "2D" indicates
#       alternating days
data = gather_data(data_codes = data_codes, 
                   start = start, 
                   end = end, 
                   freq = "Q")
```

```python
data
```

We need to align the units of Nominal and Real GDP (Billions) with the units of the quantity of base money (Millions). We will multiply the GDP measure by 1000

1 Billion = 1000 Million

```python
data["NGDP"] =  data["NGDP"].mul(1000)
data["Real GDP"] = data["Real GDP"].mul(1000)
data
```

Calculate GDP Deflator: The GDP Deflator is the  measure of the price level deduced from the calculation of real GDP.

```python
Py = data["NGDP"]
y = data["Real GDP"]

P = Py / y
#P.plot(figsize = (20,10))
P
```

```python
Py = data["NGDP"]
y = data["Real GDP"]

data["GDP Deflator"] = Py / y
data.loc["2012"]
```

Next, calculate the velocity of money. 

V = Py / M


```python
Py = data["NGDP"]
# Which measure of M??
M = data["Base: Total"]

V_base = Py / M
V_base
```

```python
Py = data["NGDP"]
# Which measure of M??
M = data["Base: Total"]

data["Base Velocity"] = Py / M
data
```

```python
Py = data["NGDP"]
# Which measure of M??
M = data["Base: Currency in Circulation"]

data["Currency Velocity"] = Py / M
data
```

How do we calculate inflation???

```python
# use periods per year = 4 if quarterly data
periods_per_year = 4
data["Inflation"] = data["GDP Deflator"].pct_change(periods_per_year)
data["1980":"1981"]
```
