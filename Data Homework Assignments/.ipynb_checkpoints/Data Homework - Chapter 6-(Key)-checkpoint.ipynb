```python
#plots.py
# if you have not installed pandas-datareader, uncomment the command 
# in the next line. After intallation, you can remove the command
#!pip install pandas-datareader
# pandas_datareader is imported in datlib.FRED
import datetime
from datlib.FRED import *
# for each variable, enter the variable name followed by the variable code. 
# the first variable is completed for you
data_codes  = {"Currency in Circulation":"MBCURRCIRW",
               "GDP Deflator":"GDPDEF",
              "NGDP": "GDP"}
# Select start and end dates
# datetime.datetime(year, month, day)
start = datetime.datetime(1990, 1, 1)
end = datetime.datetime.today()
data = gather_data(data_codes = data_codes, 
                   start = start, 
                   end = end, 
                   freq = "A")
data
```

1.	Transform Nominal GDP so that it reflects the same units as Base Money (Currency in Circulation). The final result should reflect units of Millions of Dollars instead of Billions of Dollars.

```python
data["NGDP ($ Mil)"] = data["NGDP"].mul(1000)
```

2.	Calculate Real GDP using data for Nominal GDP and the Price Level (Be sure to divide the price level by 100 first).

```python
data["GDP Deflator"] = data["GDP Deflator"].div(100)
```

```python
data["Real GDP"] = data["NGDP ($ Mil)"].div(data["GDP Deflator"])
```

3.	Calculate the Velocity of Base Money (Currency in Circulation) by dividing Nominal GDP by the values for that variable.

```python
data["Velocity"] = data["NGDP ($ Mil)"].div(data["Currency in Circulation"])
```

4.	Calculate portfolio demand for Base Money (Currency in Circulation). This is the inverse of velocity.

```python
data["k"] = data["Velocity"].pow(-1)
data
```
