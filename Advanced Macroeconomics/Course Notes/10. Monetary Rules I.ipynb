# Central Banking

The central bank is a fixture of the modern monetary environment. Central banks attempt to influence financial markets, often acting to target inflation, interest, and unemployment rates.  The theories so far presented are sufficient for interpreting the practice of central banking. Typically, the goal of the central bank is stated as promoting macroeconomic stability. In practice, that goal can be influenced by political context. Instability can result from policy decisions that are either insufficiently informed or beholden to political incentives. In this chapter we will consider the policy tools available to central banks and the dynamics that may arise in response to the policy of central banks. We begin by considering the precursor to central banking: the clearinghouse association. 

## Clearinghouse Associations

Central banks were not without precedent. They evolved much as money evolved from barter and the banking system evolved as a means of lending money from savers to borrowers. Recall the evolution of economic development presented thus far. We begin with a world of scarcity where agents own property and take action according to their preferences. As agents interact, direct (barter) exchange arises. Each agent trades some good or goods that he owns for some other good or goods that he or she values more than the item or bundle given up. In each exchange, agents pays only as much for a good or goods as they are willing. Sometimes, a good desired by an agent can only be acquired by multiple exchanges. Imagine that agent A owns an apple, agent B owns an orange, and agent C owns an avocado. Agent A is willing to trade her apple for an avocado, but agent C will only trade her avocado for an orange. It just so happens that Agent B would like to trade his orange for an apple. Agent A, sensing an opportunity to attain her desired end, trades her apple for agent B’s orange. Then she trades the orange for an avocado. This is indirect exchange. Over time, some types of commodities, say oats, come into use for indirect exchange. The commodity used for indirect exchange comes to gain value for its use in indirect exchange. This value is its exchange value. Eventually, one or several of these moneys comes to dominate markets of indirect exchange. Money has developed whose price tends to reflect its supply and demand. One interesting feature of a commodity money is that its quantity is, in the long run, dependent on its supply and demand. If money becomes dearer due to an increase in demand for it, its price will rise and thereby increase the quantity of money supplied. A fall in demand will allow the quantity supplied in a given time period to fall. Thus, the quantity of money stock is self-regulated with respect to changes in demand. 

Commodity moneys are costly for agents carry. If the commodity money is not standardized, it may be difficult to measure. If it is heavy, like gold or silver, it may be costly, or even dangerous, to carry on one’s person. Owing, at least in part, to these reasons, agents find that they benefit from leaving their money entrusted to a third party at a secure location. The agent will likely receive a deposit slip in exchange which can be used as money. Thus we have the emergence of fiduciary currency. Eventually, the agent or firm entrusted with the gold realize that they might profit from lending out some portion of the existing deposits. This allows the latent media to earn a return. This return allows the agent or firm to pay interest on deposits. The cost of holding gold is no longer borne by the agent who owns the commodity. Fractional-reserve banking is born.

Banks continually interact with one another. A account holder at one bank may wish to transfer part of his or her balance to an account holder at another bank. Every time this occurs, banks could transfer the full value of the transaction. A less costly option is to maintain an account of balances and to net out the difference between institutions at the end of each day. The clearinghouse was a banking nexus that allowed for members to perform this task. 

A clearinghouse is the location at which multiple banks may hold some of their reserves and keep their records. They have existed at different times in, for example, the United States, Canada, and England. By holding their reserves and records at a common location, banks can clear debits and credits between one another and use the reserves on hand to pay remaining debts between one another. In order to maintain their balances and ability to repay depositors, banks may sometimes borrow from one another. This need was efficiently served by other banks at the clearinghouse. This function was especially important in the case of an economic downturn.

Any bank in operation chooses to keep on hand some commodities whose value is equal to a portion of its reserves. This is needed in the case that some depositors want to immediately withdraw their currency from the bank. If too many depositors attempt to withdraw from the bank simultaneously, the bank may risk being unable to make a repayment. The bank could borrow from another bank in order to stave off hysteria, but what if many banks experience a banking panic simultaneously. If banks are unable or unwilling to lend to one another in this situation, they will have to close operations until they can acquire the funds necessary to repay depositors. This option is a last resort as it will certainly attract the attention of risk-averse depositors. 

Banks have incentive to minimize the occurrence of this scenario as a wave of collapses may have repercussions for the entire economy.  In addition to its more basic functions, a clearinghouse association helped promote stability the banking system, providing means of mitigating and even preventing an outright credit collapse.   

The clearinghouse is itself a nexus of information concerning the creditworthiness of borrowers, thus serving a role in risk mitigation. The clearinghouse might also mitigate risk by producing temporary currency during a crisis. The clearinghouse had reserves on hand that could be used to support the creation of this currency. During a crisis the clearinghouse, ostensibly drawing from these reserves, lent out emergency currency. The clearinghouse would lend to banks that it deemed to be only illiquid. Those who borrowed could repay other members of the clearinghouse system with the emergency currency. The borrowing bank was expected to repay the loan with interest once the crisis was averted.

Since credit is employed predominantly for the sake of maintaining business operations, a net decrease in available credit typically diminishes the level of future production, and consequently, real income. Likewise, a collapse in the credit markets leads to a collapse in production until the collapse reverses (assuming it reverses). A wave of banking failures, like the one that occurred during the Great Depression, can turn an economic recession into a depression. 

For the sake of self-preservation, the clearinghouse has incentive to prevent such an extreme crisis. Clearinghouses took an active role in regulating the positions of its member banks. If could, at random or on the suspicion of a poor banking practice, withdraw a large amount of funds from a member bank in order to test its capacity to handle adverse clearings. This served to discourage excessive risk taking by member banks and represented yet another means of promoting the stability within the system.  Having mechanisms that promote stability does not suggest that the system was itself perfect. Every system has bugs. There will always be some banks that take on excessive risk. There will always be failures. Irresponsible banks must fail or else they may corrupt the whole system as other banks begin to expect bailouts. The significance of the clearinghouse system is not that it prevented any instability, but that it placed bounds on that instability. Regulation and stability are themselves properties of the system as entrepreneurs earn profit by finding ways to mitigate risk.

## The Central Bank as the Lender of Last Resort

A central bank is  a financial institution either sponsored or given special privilege by the national government whose currency it regulates. The lender of last resort role is probably the strongest justification for central bank management of the money stock. During a crisis, banks need liquidity. Under some banking regimes, private banks collectively established institutions that stabilized the banking system during crises. Historically, central banks have been established because financiers and policymakers expected that the central bank could more efficiently serve this role .

The central bank is responsible, not only for providing liquidity during a crisis, but also to manage the base money stock. This is of particular significance for the functioning of credit markets. If a bank or banks risk collapse, cannot acquire credit, and appears to be solvent, the central bank’s role is to provide temporary liquidity to the institution. This function provides stability to the system during periods of credit collapse.

When credit is expanded, this is typically indicative of an increase in the nominal value of production. This can occur either through an increase in production itself, an increase in prices, or an increase in demand for cash balances which may lead to an increase in demand for credit assuming that prices do not immediately adjust to accommodate the change in demand for money. 

The relation between credit and money implies a relationship between aggregate demand and credit. When the volume of credit extended increases on net, this is equivalent to an increase in the money stock, and therefore an increase in aggregate demand (MV). Changes in the credit stock, absent a change in demand for money, shifts the aggregate demand curve in the same direction. After a period where credit has been misallocated, something only discovered ex post, there must follow a contraction as the value of collateral adjusts to reflect underlying demands of the economy. While this process works to reallocated resources to their most highly value use, the fall in credit also precipitates a fall in income. Falling incomes encourages agents and firms to take a more conservative stance financially until they expect that the crisis is nearing an end. This tendency creates aggregate oscillations known as the business cycle.

When these oscillations in the volume of available money and credit grow extreme, banks that might otherwise be economically sustainable risk default. The same holds true for other businesses. In the worst case, these fluctuations may create a wave of insolvencies that turn a minor crisis into a financial pandemic. Nominal fluctuations begin to affect the structure of the real economy. 

This occurred during the 2008-2009 crisis. Many large financial firms were holding bundles of AAA rated mortgage backed securities. The AAA rating hid from non-critical observers the risk contained in these securities. Before the crisis in 2007 and 2008, banks held these securities were able to acquire very high levels of leverage. The ratio of debt to cash on hand grew to dangerously high levels. When the value of these assets collapsed, leverage levels increased as many of these firms were on the brink of collapse. The bank's liabilities – mortgage backed securities – fell in price. This was equivalent to a drop in the broader money stock as it necessitated a contraction of credit in order for the system to stay solvent. In order to avoid collapse, the Federal Reserve provided the market with liquidity by buying up “toxic” assets, i.e., the overvalued mortgage backed securities.


# Central Banking: The Basics

Private systems of risk management represented by clearinghouse associations were largely displaced by central banks over the last one to two centuries. While the details of any particular central bank may vary,  a central bank is typically given the privilege of note issue. These notes circulate as legal tender and, as late as the first half of the 20th century, circulated alongside precious metals. Before the move to a fiat system, central bank notes were redeemable for precious metals that were supposed to back them.

Under a fiat standard, the central bank controls the quantity of base money in circulation. It may choose this quantity either directly or indirectly. Traditionally, the tools by which the central bank has affected the quantity of money were limited to three types: the discount rate, setting of reserve requirements, and open market operations.

## Discount Rate 

The central bank lends money directly to other banks through the discount window. In the early days, this was the central bank’s primary means of intervention. It has come to play a smaller role in the modern environment. In the United States, The Federal Reserve only lends to member banks. These banks therefore play an intermediate role between the Federal Reserve and the rest of the banking system.

By providing loans to banks that might otherwise fail during a crisis, the central bank intends to stabilize financial markets. When conditions improve, the borrowing banks repay the debt incurred during crisis. While this may stabilize markets in the short-run, it may promote moral hazard, leading to increased risk-taking in the long-run. 

## Open Market Operations 

This is a favorite tool of policymakers at a central bank. Open market operations is the practice whereby central banks buy and sell debt on the open market. If the central bank purchases debt, it has increased the monetary base, usually with the expectation that this increase will promote credit creation. If the central bank sells, bonds, it diminishes the base money stock and, by so doing, discourages the creation of new credit. Most central bank policies imply an inflationary bias, so this latter case is less common.

Since the creation of new money is dependent upon the creation of debt, this means that the Federal Reserve must also offset repayment when it formulates monetary policy. If the central bank purchases a bond that matures in 5 years, this means that the money that was created to purchase the bond is returned to the Federal Reserve with interest by the end of the 5 year period. Thus, if the Federal Reserve ceased making new purchases on the open market, the quantity of money would actually shrink.

The length of maturity can also play a role during a crisis. For example, Alan Greenspan, chairman of the Federal Reserve for nearly two decades, often made large purchases of short-term bonds to provide temporary liquidity if there occurred or he expected a financial crisis. This was often referred to as the "Greenspan put". This is precisely what occurred in the lead up to the year 2000, when many were worried that the banking system may experience difficulty due to a glitch that might occur due to the switch from 1999 to 2000 in the software systems. Luckily this never became a problem. Still, Greenspan prepared the market by providing temporary liquidity that was quickly removed after the start of the new year (Figure X).

When buying and selling bonds, the central bank must choose which kinds of bonds it will purchase. The central bank usually buys and sells government debt. In the U.S., this is divided into 3 categories: treasury bills, treasury notes, and treasury bonds. Treasury bills include U.S. Treasury debt that matures within one year. Treasury notes include include debt that matures in 1 to 10 years. Treasury bonds include any U.S. Treasury debt that matures over a period lasting more than 10 years. While purchases of securities by the central bank tend eventually to effect the entire financial market, the choice of maturity length will tend to move rates of the target maturity relative to other maturities. For example, if the Federal Reserve purchases securities that mature over a relatively short period, it may depress rates of that class of instruments relative to securities that mature over a longer period of time, say 30 years.

## Reserve Requirements 

The central bank can influence the broader money stock by changing the proportion of commercial bank liabilities that are required to be held with the central bank. By increasing reserve requirements, the central bank contracts the total money stock. By decreasing reserve requirements, the central bank enables the expansion of the total money stock. Reserve requirements for large financial institutions in the U.S. currently stands at 10\%. This means that the banking system, specifically deposit banking, can generate liabilities amounting to about 10 times the value of reserves in the system. 
The reserve ratio of the system of deposit banking subject to these reserve requirements is described:

>$r = \frac{R}{D}$
>
>$Where:$
>
>$r: Reserve Ratio$
>
>$R: Reserves$
>
>$D: Deposits$

The describes the reserve ratio that arises after money has been deposited at a bank. Then a fraction of the deposit, $(1 - r)$, is lent and deposited at another bank, and so forth. The money multiplier of the system can be described by the inverse of the reserve ratio, $\frac{1}{r}$.

>$M = \frac{D}{R}$

During the Great Depression, the Federal Reserve exercised this policy tool to the detriment of productivity in the United States. It 1937, the Federal Reserve chose to increase the required reserve ratio from 20\% to 35\%. The result was that the quantity of money as measure by M2, which had been growing at a steady rate, actually began to shrink. The fall in lending led to a negative aggregate demand shock the significantly reduced productivity until the policy was reversed in 1938.

## Monetary Rules and Expectations

We have spoken quite a bit about moentary policy and inflation expectations. When individuals are making decisions about how much of their income to save and it what forms this income should be saved, each must consider the expected loss or gain that will be incurred from holding money. Recall that Milton Friedman identified that expected returns on other assets play an important role in determining the velocity of money. It is significant that when one makes this decision, he or she must consider nominal rates of return. As the currency unit falls in value, nominal interest rates will tend to rise in order to reflect high inflation expectations. Thus, as inflation expectations rise, one can expect that portfolio demand for money will fall. This dynamic has served as the basis for much of our discussion of inflation and inflation expectations.

The targeting of inflation is a consequence of a desire for price stability. The dysfunction that we have observed in financial markets during major crises has encouraged an inflationary bias among monetary theorists. Under a regime with a modest, positive rate of inflation, central banks can, with relative ease, lower target interest rates without running into the zero-lower-bound problem. Central bankers aim to avoid deflation of any kind and elevated inflation. To the extent that policymakers error, the bias tends to be in the direction of inflation, as we will see in the data.

In our discussion of recent monetary history, a number of theoretical concepts were introduced that we will continue to unpack. We have focused largely on the relationship between monetary policy and expectations, with primary focus on the relationship of monetary policy to inflation. Yet, we also know that monetary policy can have short-run impacts on real income. In fact, if real income lacked a cyclical component all together, there would be little justification for monetary management. The central bank, established as the lender of last resort, is expected to provide stabilzing intervention in the face of a significant economic downturn. 

### Rules vs. Discretion

Rising inflation during the 1970s motivated the study and discussion of monetary rules. High inflation appeared to be the result of a lax policy. As we have already learned, inflation was high and volatile because inflation expectations had come unanchored. The goal of monetary policy was unclear. It was likely motivated in part by a desire to support public spending and also a desire to avoid overseeing economic contraction coordinated by monetary policy. It was guided, similarly, by a belief that higher inflation would reduce unemployment. The macroeconomics of the 1980s was a revolution against this destabilizing view.

Optimal policy was shown to be, in fact, non-optimal. That is, moentary policy aimed at fooling consumer, investor, and labor to expend more resources than they otherwise would in order to lift real income in the short-run, would generate expectations of inflationary policy in the long-run. While we have learned about the short-run long-run distnction, it is worth noting that, in line with Lucas (1971), countries with relatively high and volatile inflation tend to expereince relatively limited short-run effects from moentary policy. Instead, individuals integrate inflation into their expectations relatively quickly, adjusting their investment decisions and demand to hold cash balances. Thus, attempts to optimize policy by generating unexpected positive aggregate demand shocks are self-defeating. 

The optimal monetary policy, on the other hand, clarifies _how_ the central bank will respond to changes in the economic environment. Monetary policy becomes rule bound. Rules clarify which economic variables guide policy and which variables are targeted by policy. In order to determine the appropriate rule, we must first clarify, theoretically, what variables are valuable indicators of macroeconomic activity and which are most significant in influencing those indicators. In 1977, the Federal Reserve was mandated to promote 1) maximum employment, 2) low and stable inflation (price level stability), and 3) interest rate stability. Thus, when Paul Volcker assumed leadership at the Federal Reserve in 1979, his task was straight forward to define: lower and stabilize inflation. The attitude of Volcker and most involved in monetary policy is that, under conditions of high inflation, stabilizing and lowering the rate of inflation is the best means of maximizing employment.

During Volcker period, we saw experimentation with targeting of higher monetary aggregates as well as an interest rate target that simultaneously targeted the level of reserves in the financial system. Volcker's policy was rule-like, in that it appeared to be guided by a rule that was implied by the Federal Reserve's goals and a consistent means of interpreting economic conditions and responding to them in pursuit of this goal.

Conceptually, however, our description of the rule-based policy has been opaque. When should monetary policy loosen or tighten, and by how much? This question is actually a topic of significant controversy in macroeconomics. As we can already see, central bankers must choose which variables to target as well as the variables that influence or determine their adjustment of the target variable. In what follows, we will review some monetary rules and interpret them using data employed by these rules.

### Reaction Function
        
Monetary policy might be strictly defined by an explicit or an implicit rule. When we attempt to interpret monetary policy by a rule, we refer to this rule as a reaction function. Monetary policy usually allows for a fair amount of flexibility in implementation, so, in practice, we can think of monetary rules as being loosely defined. Woodford states this concept in rather extreme form, "the [central] bank should consider what an optimal rule of conduct is . . . [and] reconsider this question as often as it likes." In reality, responsible policymakers typically maintain stable policy regimes, changing operations only when if the change has been carefully considered and vetted or in the case that extreme conditions appear to justify a change. When the policy regime does transform, it is important for policymakers to act in a manner that is consistent with policy goals so as to maintain policy credibility.

### Interest Rate Targeting 

Most rule-based policy targets interest rates. However, the mechanism that guides the setting of interest rates is critical. Quite often, this part of the response is not included in the monetary rule. Yet, the means by which a variable is targeted will likely impact economic outcomes. I wish not to linger on this point, but it is important to recognize. When we review the modern monetary policy regime, often referred to as "Unconventional Monetary Policy" (though it is conventional by now), we will see that the interest rate target is implemented by setting of payment of interest on excess reserves rather than only by adjustment of the level of reserve available in the financial system. 

Regardless of the mechanism of implementation, we can use macroeconomic data to generate *ex post* estimates of the target rate suggested by a monetary rule. We can then compare the observed values of the variable that would be targeted to see if the outcome predicted by the rule is similar to the outcome that manifested.

### Wicksell Rule

The Wicksell rule calls for an increase in interest rates in order to offset changes in the price level.

$$p_t = E(p_{t+1}) + r_t - i_t$$
$$ i_t = \bar{i}_t + \phi p_t$$
(p. 50)

Can be simplified:

$$\Delta i_t = \pi_t$$


```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
def gather_data(data_codes, start, end = datetime.datetime.today(), freq = "A"):
    i = 0
    # dct.items() calls key and value that key points to
    for key, val in data_codes.items():
        if i == 0:
            # Create dataframe for first variable, then rename column
            df = web.DataReader(val, "fred", start, end).resample(freq).mean()
            df.rename(columns = {val:key}, inplace = True) 
            i = None
        else:
            # If dataframe already exists, add new column
            df[key] = web.DataReader(val, "fred", start, end).resample(freq).mean()
    return df
```

```python
data_dict = {"$P_{CPI}$":"CPIAUCSL",
             "Y":"GDP",
             "y":"GDPC1",
             "$y^*$":"GDPPOT",
             "Total Base": "BOGMBASE",
             "C":"WCURCIR",
             "Reserves at Fed":"RESBALNS"}

rates_dict ={"FFR": "DFF",
             "$r_{3M}$":"TB3MS",
             "$r_{1}$": "DGS1",
             "$r_{10}$": "DGS10",
             "$r_{30}$": "DGS30",
             "U":"UNRATE",
             "$U^*$":"NROU"}

diff = 12
freq = "M"
start = datetime.datetime(1950, 1, 1)
# end = datetime.datetime(2019, 12, 31)
end =datetime.datetime.today()
data = {}
data["Level"] = gather_data(data_dict, start, 
          end = end, freq = freq)
data["Level"]["$P_{Deflator}$"] = data["Level"]["Y"].div(data["Level"]["y"]) 
data["Level"]["C"] = data["Level"]["Total Base"].div(1000).sub(data["Level"]["Reserves at Fed"])
data["Level"]["$\\frac{C}{P}$"] = data["Level"]["C"].div(data["Level"]["$P_{CPI}$"])
for y in ["Y", "y","$y^*$", "$P_{Deflator}$"]:
    # fill in missing monthly observations since data is quarterly
    data["Level"][y] = data["Level"][y].interpolate(method='polynomial', order=3)  


rates =  gather_data(rates_dict, start, 
          end = end, freq = freq)
rates["$U^*$"] = rates["$U^*$"].interpolate(method='polynomial', order=3)
for key in rates:
    rates[key] = rates[key].div(100)
    
data["Log"] = np.log(data["Level"])
data["Rates"] = data["Log"].diff(diff)

for rate in rates:
    data["Rates"][rate] = rates[rate]
    data["Log"][rate] = rates[rate]
data["Rates"].rename(columns = {"$P_{CPI}$":"$\pi_{CPI}$"}, inplace = True)
data["Rates"]
```

### Loss Function Components

It is more common now to target rates of different variables. A typical inflation target, for example, is 2%. This is the target for the Federal Reserve. Monetary policy may consider the difference between observed inflation and the target interest rate, shifting policy in light of this discrepancy. Likewise, monetary policy may attempt to "target" potential GDP or the natural rate of unemployment. These rules aim to stabilize real productivity at a rate of growth that is sustainable. (Note that we use logged values of $y$ in order to approximate the proportion of deviation).

$$\pi_t - \pi^*$$

$$y_t - y^*$$

$$U_t - U^*$$

In many cases, a rule will square such differencfes in order to penalize deviations:

$$(\pi_t - \pi^*)^2$$

$$(y_t - y^*)^2$$

$$(U_t - U^*)^2$$

In order to maintain the direction of magnitude of deviation from target (i.e., indicate that observation is above or below), we will maintain the positive or negative sign from the non-squared loss components.

To summarize, briefly:
$$L_\pi = (\pi_t - \pi_{target})$$
$$L_U = (U_t - U_{target})$$
$$U_{target} = U_{N_t}$$

One might interpret the difference of these values as indicating a linear loss function:

$$L = L_\pi - L_U$$

Negative values of L would merit monetary support whereas positive values of L, or perhaps smaller negative values, would merit diminishment of the size of the balance sheet. Likewise, deviations from target are penalized in proportion to the size of the deviation. 

Typically the loss function is presented as the sum of squared values of its components. We will square the values of the loss function components while maintaining positive and negative magnitudes of the untransformed components, thus augmenting the policy response in light of the magnitude deviations without obscuring their direction of changes to policy instruments. 

$$L_\pi^` =\begin{cases}
{L_\pi}^2 \text{ if }L_\pi > 0 
&& 0 \text{ if }L_\pi = 0
&& -{L_\pi}^2 \text{ if }L_\pi < 0
    \end{cases}$$ 

$$L_U^` =\begin{cases}
{L_U}^2 \text{ if }L_U > 0 
&& 0 \text{ if }L_U = 0
&& -{L_U}^2 \text{ if }L_U < 0
    \end{cases} $$
   

These are used to construct an augmented loss function.
$$L^` = {L_\pi}^` - {L_U}^`$$

```python
inflation_target = .02
loss_function_components = {}
loss_function_components["($\\pi_t - \\pi^*$)"] = data["Rates"]["$\pi_{CPI}$"].sub(.02)
loss_function_components["($y_t - y^*$)"] = data["Log"]["y"].sub(data["Log"]["$y^*$"])
loss_function_components["($U_t - U^*$)"] = data["Rates"]["U"].sub(data["Rates"]["$U^*$"])
# loss_function_components["($r_t - r^*$)"] = data["Rates"]["FFR"].sub(data["Rates"]["$r^*$"])

loss_function_components = pd.DataFrame(loss_function_components)
for key, val in loss_function_components.items():
    loss_function_components[key + "$^2$"] = val.pow(2)
loss_function_components
```

### Taylor Rule
Slightly different than Taylor's formulation, Woodford does not assert a growth rate of real GDP. We will use an estimate of potential real GDP.

$$i_t = \alpha + \gamma (\bar{\pi}_t - \pi^*) + \lambda (y_t - y_t^*)$$

Woodford uses:

$$i_t = 0.04 + 1.5 (\bar{\pi}_t - \pi^*) + .5 (y_t - y_t^*)$$



```python
data["Rates"]["Taylor Rule"] = 0.04 + 1.5 * loss_function_components["($\\pi_t - \\pi^*$)"] + .5 * loss_function_components["($y_t - y^*$)"]
plot_data = data["Rates"]
x,y = "Taylor Rule", "FFR"
fig, ax = plt.subplots(figsize=(20,10))
ax.axhline(0, ls = "--", linewidth = 3, color = "k", alpha = .3)
plot_data[[x,y]].plot(linewidth = 3, ax = ax)
ax.set_title("$r^2: " + plot_data[[x,y]].corr().pow(2).round(2).astype(str)[x][y] + "$")
ax.legend()
```

### Modeling the Fed's Dual Mondate

As noted before, the Fed is mandated to mantain 1) maximum employment, 2) low and stable inflation (price level stability), and 3) interest rate stability. It concentrates on 1) and 2) to maintain 3). We can model this mandate by subtracting the squared magnitude of the loss function components with the original direction of magnitude of each component maintained. Negative values would suggest a need for monetary easing while positive components suggest a need for tightening. As we can see, stability in this function emrged under Volcker and Greenspan. In the last 15 years, we have, however, epecerienced significant instability.

```python
# import mpl_axes_aligner
data["Rates"]["Fed Mandate"] = L_prime["($\\pi_t - \\pi^*$)$^2$"].sub(L_prime["($U_t - U^*$)$^2$"])
plot_df = data["Rates"].copy()
fig, ax = plt.subplots(figsize=(20,10))
ax.axhline(0, ls = "--", linewidth = 4, color = "k", alpha = .3)
plot_df[["Fed Mandate"]].plot(linewidth = 4, ax = ax)
ax.axvline("1983", linewidth = 5, ls = "-", color = "k", alpha = .5)

ax.set_title("Maintain Direction of Magnitude\nfor Squared Loss Components")
```
