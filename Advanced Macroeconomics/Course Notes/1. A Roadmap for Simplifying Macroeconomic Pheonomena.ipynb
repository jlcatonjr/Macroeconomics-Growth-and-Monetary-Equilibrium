## 1.1 Motivation

It is a common habit among macroeconomists, and academics in general, to treat undergraduate students as mini-graduate students. And it is common among teachers of graduate macroeconomics to provide a constellation of models whose mutual relevance is unclear. Often, subject matter is taught without regard to internal consistency between models or, failing that, without concern for external validity, that is, consistency between the assumptions of the models taught and the world as it actually exists. Chief among these are assumptions of perfect information, symmetrical interpretations between agents, and perfectly competitive markets. These features of macroeconomics, which are most unfortunate, have ironically become even most problematic as a result of the search for microeconomic foundations that accompanied the Rational Expectations Revolution. 

The Financial Crisis of 2008 shook the faith of the public and of many practicing economists in macroeconomic theory. Financial crises are complex issues that occur because they surprise general sentiment. Our field of study was accused of using models that do not allow for  "Black Swan" events, those crises that cannot be predicted by patterns that we have observed in the past. Part of this shortcoming of macroeconomic models was a function of its unrealistic assumptions capable only of  describing the world of the familiar. Uncertainty not accounted for by these models seemed sufficient reason to doubt the foundations of macroeconomics, although these foundations are not subject to the same critique. In the midst of hysteria, the details of macroeconomic theory and monetary policy implementation can be, and in this case were, not given proper attention.

Macreconomic theory as it is understood today is largely a descendant of the rational expectations revolution and its later derivative, the efficient markets hypothesis. For agents to have *rational* expectations is to assert that they form expectations in light of all available information. Since the theory asserts that agents are right on average, prices always reflect the full set of available information available to those agents. The efficient markets hypothesis propose that prices in financial markets integrate all available information, dividing the kinds of information in light of the strength of this assertion. In its weak form, it indicates that all information from past price data is integrated into the present price. The stronger the form, though, the more general the claim concerning integration of available information. Once integrated, any further change in prices is simply due to randomness that can be defined as changes drawn from a normal distribution of price changes centered upon a mean of 0%. In the case that prices follow a long term trend, then prices follow a random walk with drift. The distribution of price changes is centered around a value different than 0%. The mean of the distribution is positive in the case that the trend is positive and negative in the case that the trend is negative. We will focus on cases where the mean of the distribution is 0\%.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.size': 30})
plt.rcParams['axes.xmargin'] = .001
plt.rcParams['axes.ymargin'] = .005

def normal_dist(x , mean , sd):
    prob_density = 1 / (sd * (2 * np.pi) ** .5) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
fig, ax = plt.subplots(figsize = (20, 10))
x = np.linspace(-4, 4, 1000)
normal_dist_list = normal_dist(x, mean=0, sd=1)
ax.plot(x, normal_dist_list, c='C0', linewidth = 3)
ax.set_xlabel("$\Delta P$")
ax.set_ylabel("Density")
ax.set_yticklabels([str(int(y * 100)) + "%" for y in ax.get_yticks()])
ax.set_xticklabels([str(int(i)) + "%" for i in ax.get_xticks()])
ax.axvline(0, linewidth = 2, color = "k", ls = "--", alpha = .75)
ax.set_title("Convergent Expecations in Static Equilibrium")
```

The problem with an economic crisis is that it would not happen if, on average, everyone expected it. If everyone did expect it, they would hedge for it and, by virtue of the expectation, prevent a downturn from happening. Expectations concerning the future value of stocks, on the other hand, are often subject to information cascades. Some bear in the market has a hunch and, over time, the bear's narrative is convincing to other, perhaps because the generation of new facts seems to consistently align with the bear's narrative. As sentiment transforms to match the bears sentiment, expectations have shifted. Before this occurs, there exists divergent expectations (i.e., see Lachmann on Expectations). As the bear's narrative attracts attention, the set of agents who act according to the perspective go from representing the left tail of the curve representing expectations of price changes to forming the left portion of a bi-modal distribution. 

Below I represent a bimodal distribution evenly weighted betwee bulls and bears. Like the random-normal distribution, the overall mean of the bimodal distribution is 0\%, but the path of future prices is subject to much greater volatility. While the structure of the path of prices may seem to be the same, notice that the vertical axis of the prices generated using the bimodal distribution is much greater. This, of course, is not a fully representative case, since the shape and mean of distributions are constantly in flux in light of cascade effects, but it does provide a starting point for think about problems that may be generated from modeling assumptions.

### 1.2.1 The Economy in a Single Model: The Equation of Exchange

One feature that goes back at least to David Hume is the quantity theory of money and the more general equation of exchange. The equation of exchange is simply an identity. It identifies that, in equilibrium, the value of expenditures must equal the value of goods produced.

### $$MV \equiv PY$$

<center>
    
### $M$: Quantity of Money

### $V$: Average Velocity (Rate of Expenditure) of a Unit of Money

### $P$: Average Level of Prices (Price Level)

### $y$: Real Income
</center>

We divide macroeconomic activity between total expenditures, $MV$ and the total value of goods produced $Py$. We will often refer to values of these variables at a particular time as $M_tV_t\equiv P_ty_t$. In equilibrium, markets clear and all goods are purchased. And even out of equilibrium, it recognizes that the total value of expenditures is equal to the value of actual goods purchased, even if not all goods produced are, in fact, purchased. That is a situation where real expenditures are less than the value of real income produced, $y_0$:

$$\frac{M_tV_t}{P_t} < y_0$$

$$y_t < y_0$$

Thus, the equation of exchange helps us to recognize that there may be cases where the economy is producing real income at levels, or rates, significantly below its potential. It allows us to ask questions concerning factors that impact immediate productivity as well as productivity over long periods of time.

This may seem like an especially elementary concept. *It is!* And without the Equation of Exchange, we would be unable to sensibly discuss economic activity across the entire economic system. While there are many details that the equation of exchange leaves out, it is a tool that we cannot do without. It is this concept that we will leverage throughout our investigation of the macroeconomy as well as controversies that have arisen during the course of development of macroeconomic theory.

### 1.2.1.1 The Quantity Theory of Money and the Stability of Money Demand

Many controversies in macroeconomic theory concern the stability of money demand. As we will see, the John Maynard Keynes asserted that demand to hold money tends to be unstable and that, under conditions of extreme pessimism, individuals will only increase their cash balances as new money is created.  The quantity theory asserts that the average velocity of money and real income are constant. Therefore changes in the quantity of money are highly correlated with inflation:

$$M\bar{V} = P\bar{y}$$

Recall that for $y=log(x)$: 

$$\frac{dy}{dx} = \frac{1}{x}$$

Therefore:
$$dy = \frac{dx}{x}$$

The derivative of a log is the percent change of change of x. The difference of logged values over time is analogous to the derivative of the log. This distance over a discrete period of time approximates the rate of change. If we log and difference the variables, we find that:

$$\%\Delta M \approx \%\Delta P$$

since:
<center>
$$\%\Delta V = 0$$
$$\%\Delta y = 0$$
</center>

This assertion is a simplified means of accounting for the fact that demand for money is stable. Here, $V$ is the inverse of demand to hold money, often referred to as portfolio demand for money, $k$, and $y$ that refers to transaction demand for money. As we investigate monetarism, we will find that a more realistic approach anchors $k$ using the interest rate and, similarly, $y$ with the return to capital. These are actually the same thing in an equilibrium with no transaction costs. Therefore, it is sufficent to recognize that demand for money is stabilized by the interest rate. 

$$V_t = f(i_t)$$


```python
# !pip install pandas_datareader
import datetime
import pandas as pd
import pandas_datareader.data as web

def gather_data(data_codes, start, end = datetime.datetime.today(), freq = "A"):
    i = 0
    # dct.items() calls key and value that key points to
    for key, val in data_codes.items():
        if i == 0:
            # Create dataframe for first variable, then rename column
            df = web.DataReader(val, "fred", start, end).resample(freq).mean()
            df.rename(columns = {val:key}, inplace = True) 
            i = None
        else:
            # If dataframe already exists, add new column
            df[key] = web.DataReader(val, "fred", start, end).resample(freq).mean()
            
    return df
```

I bring this relatively advanced fact into early discussion as the stability of money velocity continues to be a contentious topic in spite of the significant intellectual and political victories of monetarism during the high and volatile inflation of the 1970s. While, owing to his emphasis of the Quantity Theory, Milton Friedman's monetarism has come to represent a shorthand reference to the theory, Friedman's elaboration of the quantity theory in 1956 recognized the tradeoff between holding cash and other assets in light of the expected returns of each. In other words, demand to hold money was a function of rates of return. Thus, money demand is often accounted for by inclusion of a _nominal_ interest rate. In other words, velocity can be analytically anchored by a generally representative interest rate. As we will see, there is significant relationship between the rate of expansion of the quantity of money, expected inflation, and interest rates. This fact will link the equation of exchange to financial markets. Friedman (1971) was quite explicit about this connection.

The changes in nominal interest rates are also strongly correlated with with changes in nominal income, thus further tying the equation of exchange to financial markets. This should be no surprise. Nominal income can divided into to variables, the price level $P$ and the real income, $y$. In dynamic analysis, we refer to the rate of change of $P$ as inflation, $\pi$. Thus, the rate of growth of real income can be divided between inflation and real income: 

$$\% \Delta Py \approx \pi + \% \Delta y$$

Interest rates, on the other hand, correlate with expected income growth. In macroeconomics, we assert that we are in equilibrium when expectations of values match realized values. According to the Fisher Equation, interest rates contain a nominal, in the form of the expected rate of inflation, and a real component (We could, of course, add additional components like risk premia which themselves can be further subdivided.):

### $$i_t = r_t + \pi_{e_t}$$

Like the relationship between velocity and the interest rate, the correlation between nominal interest rates and the growth rate of nominal income also tend to hold true. 

While it is tempting to read interest rates out of monetarist analysis, as an uncareful reading of Friedman would suggest, there is no need to import such an emphasis to the present time where monetarist analysis that excludes interest rates, including interest rate targeting and expectations, is a sure path to drawing nonsensical inferences.


### 1.2.2 Expectations: Where the Present Meets the Future

While rational expectations has been subject to critique in recent years, the shift of focus by macroeconomists to expectations has had profound impact both on theory and policy. In thinking about expectations, we need not assert that the crowd is always correct on average, though often investor expectations represent the best interpretation of present conditions and their impact on the future. The expectations revolution recognizes that individuals often expect the near future to look like the recent past. This sort of view is known as *adaptive expectations*, whereas the move to emphasizing the ability of markets to integrate presently available information in order to form expectations about the future is consistent with *rational expectations*. Yet, these are not an exhaustive list of theoretical approaches to expectations. 

Our interpretation of the word, *expectations*, should depend not on a specific theoretical form but, rather, the simple fact that the present actions of individuals are always formed in light of future conditions expected by those individuals. With this in mind, we can recognize empirical facts related to expectation formation and action. For example, when individuals feel less certain about the future, on average, they will likely expend more resources safeguarding their wealth so that they are better off than they would otherwise be if less than ideal economics conditions manifest. 

Since they impact preferences that motivate present action, expectations of the future drive present changes in both the price level and relative prices. Changes in the price level not justified by past expansion of the quantity of money imply rising inflation expectations. That is, consumers and investors expect that prices will rise in the future due to an expansion of the quantity of money, they will often act in a manner that promotes the price level increase before monetary expansion occurs. 
Likewise, monetary expansion that is expected to reverse in the foreseeable future will have little to no effect on present prices. 

Such ineffective expansions are more common than one might expect. For example, Alan Greenspan greatly expanded the quantity of base money in the build up to the year 2000, but then allowed the quantity of money to return to trend after the expansion. This was intended to quell fears around *Y2K*, which some believed would lead to errors in various accounting systems. While some blame the expansion for a temporary overvaluation of equities - i.e., the NASDAQ bubble - even if the claim has merit, the policy succeeded in calming markets at the time and, at worst, generated inflation consistent with the much of the rates seen earlier in the decade. Likewise, the Bank of Japan attempted to motivate an increase in expenditures by lowering interest rates, however a commitment to limiting inflation led investors to be skeptical of the willingness of the Bank of Japan to allow for inflation. [Without significant and permanent monetary expansion](web.mit.edu/krugman/www/japtrap.html), a policy of easing failed to be effective due to expectation of tightening that would result if the expansion was, in fact effective. As Sumner points out, eventually the Bank of Japan did significantly expand currency, but it was not committed maintaining this expansion. Investor confidence in the central bank to be responsible in limiting inflation translated into a curse of competence.

### 1.2.3 Technology and Growth

Finally, we will also explicate the factors that influence economic growth. While this will be the first concept that we dissect, I emphasize this last in this list as economic growth refers to growth of real income, which we have been situating so far within a framework of aggregate analysis. Our approach to growth theory will intentionally concentrate on the Solow model, which framed the development of growth theory that followed, as well as Real Business Cycle Theory, which identifies a emphasis on economic growth as the logical consequence of adoption of new classical methodology with its assumptions concerning agent knowledge and market efficiency.

The Solow model will identify the key fact of growth theory, which is that capital accumulation on its own cannot promote economic growth. Rather, technological advancement promotes growth of real income per capita. 

$$y_t = f(A, k_t)$$

$A$: Technical Efficiency of Capital

$k_t$: Capital per Unit of Labor ($\frac{K_t}{L_t}$)

The model is simplistic, but that is okay. If interpreted literally, suggests that savings and capital accumulation do not influence technological development. Further, A is a black box, meaning that it accounts for all factors not identified by *k*. That might make the model useless. It was intended by Solow to bea model of economic growth. Yet, the very thing that is fails to explain it at least identifies as the source of economic growth. Much of the growth theory that followed attempts to define $A$ and to explain how the value of $A$ advances so as to promote economic growth. Paul Romer, for example, attempts to endogenous innovation by firms by identifying efforts at innovation as a function of profits that result from assymetry of knowledge and ability by firms. Romer also identified learning that occurs in economic exchange as a source another source of technological advancement. Others, like Daron Acemoglu and Douglass North have built careers showing how good institutions lower transaction costs and, as a result, increase in economic efficiency that facilitates economic development.


## 1.3 The Power of Policy to Obscure Interpretation

Often, empirical data will appear to be at odds with theoretical assertions. This is often due to errors in interpretation. For example, Paul Samuelson championed growth in the U.S.S.R. when, in reality, economic data provided by Soviet and eastern block countries where economic data was distorted as there existed little in the way of democratic oversight of data collection. In this case, the data was bias. Sometimes, the data is just more complicated than one would expect. Earlier I mentioned that a naive employment of the quantity theory can lead to just such a view. It appears, by a cursory view of the data, that the quantity theory no longer describes monetary affairs. In fact, to the extent that inflation and the rate of growth of circulating currency are related, this relationship is negative. A more careful, but still incorrect, interpretation is that there is no economically significant relationship between inflation and the rate of growth of currency. Just a modest background in statistics should cue one to be skeptical of this view. Simple correlation between variables whose impacts are felt across time and are potentially endogenous will often be incorrect. In the least, we must control for the relevant variables, indicated by the equation of exchange and its expansion in the form of a monetary theory of nominal income. 

The quantity theory appears to hold during a period where the rate of monetary expansion did not greatly vary and monetary policy was not set with intention of coordinating investor expectations. Ability of policy to obscure theory does not stop there. In 2008, Benjamin Bernanke began to implement policy that at the time was referred to as *unconventional moentary policy*. By now, this policy has become conventional. Instead of targeting interest rates by adjusting the rate of accumulation of U.S. Treasuries, the Bernanke Fed more than doubled the value of its assets while simultaneoulsy borrowing reserves in order to prevent inflation. This accounting trick made it seem as though monetary expansion was no longer effective in lifting expenditures. Many who awaited hyperinflation had their expectations upset. A rush to hold tangible assets like gold and silver from those critical of the new regime ultimately was upset. Precious metals markets experienced a bubble that had peaked and deflated by 2011. 

To make interpretation of monetary policy more difficult, the Federal Reserve expanded operations on multiple margins. It began to hold mortgage backed securities, lending to other central banks, and borrowing significantly from the overnight lending market. Without a theory to understand *how* monetary policy affects economic outcomes, the new operations appear to be a complicated mess that turn moentary theory in its head. As [Benjamin Bernanke said](https://youtu.be/5cTFE3BbGv4?t=1416) in an discussion at Brookings Institution:

> "The problem with QE is that it works in practice, but it doesn't work in theory."

However, we will see, again with our discussion of a monetary theory of nominal income, that monetary policy still operates according to the basic principles outlined here and that will be further elaborated over the course of the text. We will see that monetary dynamics with regard to inflation and interest rates still follow the same principles even if interest rates are set using a mechanism different than monetary expansion via open market operations. Good analysis simply requires that one be critical when selecting data and careful with the theory employed to analyze it.
