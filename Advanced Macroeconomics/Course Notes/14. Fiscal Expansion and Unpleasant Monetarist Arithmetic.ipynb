## Thesis: "The purpose of this paper is to argue that, even in an economy that satisfies moentarist assumptions, if monetary policy is interpreted as open market operations, then Friedman's list of the things that monetary policy cannot permanently control may have to be expanded to include inflation."

"The public's demand for interest-bearing government debt constrains the government of a monetarist economy in at least two ways. (For simplicity, we will refer to publicly held interest-bearing government debt as _government bonds_.) One way the public's demand for bonds constrains government is by setting an upper limit on the real stock f government bonds relative to the size of the economy. Another way is by affecting the interest ratethe government must pay on bonds."

1) If monetary policy dominates fiscal policy, then "[t]he fiscal authority then faces the constraints imposed by the demands for bonds, since it must set its budgets so that any deficits can be financed by a combination of the signorage chosen by the monetary authority and bond sales to the public." As a result, the monetary authority can control inflation.

2) If fiscal policy dominates monetary policy, the fiscal authority sets its budget and monetary policy must respond to support the aims of fiscal policy. "If the fiscal authority's deficits cannot be financed solely by new bond sales, then the monetary authority is forced to create money and tolerate additional inflation. . . . [s]uppose that the demand for government bonds implies an interest rate on bonds greater than the economy's rate of growth. Then, if the fiscal authority runs deficits, the monetary authority is unable to control either the growth rate of the monetary base or inflation forever. . . . Being limited simply to dividing government debt between bonds and base money and getting no help from budget surpluses, a moneary authority trying to fight current inflation can onlly do so by holding down the growth of base money and letting the real stock of bonds held by the public grow. If the principal and interest due on these additional bonds are raised by selling still more bonds, so as to continue to hold down the growth in base money, then because the interest rate on bonds is greater than the economy's growth rate, the real stock of bonds will grow faster than the size of the economy. This cannot go on forever, since the demand for bonds places an upper limit on the stock of bonds relative to the size of the economy. Once that limit is reached, the principal and interest due on the bonds already sold to fight inflation must be financed, at least in part, by seignorage, requiring the creation of additoinal base money."

## Tighter money now can mean higher inflation eventually

Assumptions

a. A common constant growth rate of _n_ for real income and population

b. A constant real return on government securities that exceeds _n_

c. A quantity theory demand schedule for base or high-powered money, one that exhibits constant income velocity



#### The real deficit is the increase in the real stock of money (purchasing power for the government) from new base money created in period t. We add to this, new borrowing and subtract real interest payments (the matured value of old borrowing from the previous period).

1) $D(t) = \frac{M_B(t) - M_B(t-1)}{p(t)} + B(t) - B(t-1)[1 + R(t-1)]$

Where:

> $H(t)$: High-powered (Base) money at time $t$
>
> $B(t)$: Real value of bonds held by public
>
> $R(t)$: Real interest rate on bonds

May be rewritten

1.1)  $D(t) = \frac{\Delta M_B(t)}{p(t)} + B(t) - B(t-1)[1 + R(t-1)]$

#### Population grows at a constant rate, $n$
2) $N(t+1) = (1+n)N(t)$

Where:

> $N(t)$: Population at time $t$
>
> $n$: Growth rate of population

#### Bonds per capita at time $t$ are augmented by the interest and moderated by population growth. We add new real deficit less the portion of the deficit supported by money creation.

3) $\frac{B(t)}{N(t)} = \frac{1 + R(t-1)}{1+n}\frac{B(t-1)}{N(t-1)} + \frac{D(t)}{N(t)} - \frac{M_B(t)-M_B(t-1)}{N(t)p(t)}$


Where:
>$\frac{B(t)}{N(t)}$: Real value of bonds per capita

#### $M_B(1)$ is given. $\theta$ is chosen by monetary authority. Key to assumption that monetary policy dominates fiscal policy

4) $M_B(t) = (1 + \theta)M_B(t-1)$

#### The price level in the present period is a function of High-powered money per capita (presumes there is some constant level of per capita real income)

5) $p(t) = \frac{1}{h}\frac{M_B(t)}{N(t)}$

Where:

> $h > 1$, therefore $\frac{1}{h} < 1$

#### The growing population necessarily increases demand for real cash balances since agents are identical. Inflation will be a function of the per capita deficit and its growth path which is determind by the real value of bonds per capita each period.

6) $1 - \frac{ \frac{p(t-1)}{p(t)}}{1+n} = \frac{D(t)}{N(t)} + \frac{R(t-1)-n}{1+n}\frac{b_\theta(t)}{h}$

Where:
>$b_{\theta}(t) = \frac{B(t)}{N(t)}$

#### Less monetary expansion implies greater dependence on bonds to finance deficit. So a lower rate of monetary expansion can imply a higher rate of indebtedness

7) $b(1) = \frac{\bar{B}(0)}{N(1)p(1)} + \frac{D(1)}{N(1)} - \frac{M_B(1)-M_B(0)}{N(1)p(1)}$

8) $b(t) = \frac{1 + R(t-1)}{1+n}b(t-1) + \frac{D(t)}{N(t)} - \frac{h\theta}{1+\theta}$

#### The larger is $b_\theta(t)$ is the consequence of smaller is $\theta$

9) $b_\theta(t) = \phi(t,1)b(1) + \sum_{s=2}^{t}{\phi(t,s)\frac{D(s)}{N(s)}} - \frac{h\theta}{1+\theta} \sum_{s=2}^{t}{\phi(t,s)}$

9.1) where $\phi(t,t) = 1$ for $t>s$: 

$$\phi(t,s) = \frac{\Pi_{j=s}^{t-1}{(1 + R(j))}}{1+n}^{t-s}$$

It is crucial for such a result that the real rate of return on government securities exceed $n$ from $T$ onward (6) and that the path of fiscal policy given by D(1), D(2), . . ., D(t), . . . not depend on $\theta$."

## Tight money now can mean high inflation now

#### If fiscal policy dominates monetary policy, then investors will expect that government debt may grow faster than real income, which in this case is growing alongside the population. The price level here is not just a function of the present money stock, but upon expectations of the future money stock. If the government lacks income to repay its debts, since debt is growing faster than real income, than it must necessarily default, either explicitly, or implicitly through inflation. 

#### In an environment where fiscal policy is insensitive to the goals of monetary policy, if the central bank increases the interest rate target in order to avoid this scenario, investors are aware that the fiscal authorities will eventually win this fight. In this case, the value of $\theta$, governing the rate of monetary expansion, will tend to be higher.

#### "If the demand for money depends on the expected rate of inflation, then it turns out (see Sargent and Wallace 1973) that the current price level depends on the current level and all anticipated future levels of the money supply. This sets up a force whereby high rates of money creation anticipated in the future tend to raise the current rate of inflation. As we shall show, this force can limit the power of tighter monetary policy to deliver even a temporarily lower inflation rate."

Replace equation 5) for money demand equation:

10) $\frac{M_B(t)}{N(t)p(t)} = \frac{\gamma_1}{2} - \frac{\gamma_2}{2}\frac{p(t+1)}{p(t)}$

$\gamma_1 > \gamma_2 > 0$

#### The price level is a function of the (expected) path of future money growth.

10.1) $p(t) = \frac{2}{\gamma_1} \sum_{j=0}^{\infty}{(\frac{\gamma_2}{\gamma_1})^j\frac{M_B(t+j)}{N(t+j)}}$



```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import datetime
import statsmodels.formula.api as smf
import statsmodels.api as sm
import datetime
import warnings 
warnings.filterwarnings("ignore")

def gather_data(data_codes, start, end = datetime.datetime.today(), freq = "A"):
    i = 0
    # dct.items() calls key and value that key points to
    for key, val in data_codes.items():
        if i == 0:
            # Create dataframe for first variable, then rename column
            df = web.DataReader(val, "fred", start, end).resample(freq).mean()
            df.rename(columns = {val:key}, inplace = True) 
            i = None
        else:
            # If dataframe already exists, add new column
            df[key] = web.DataReader(val, "fred", start, end).resample(freq).mean()

    return df

data_codes = {"CPI": "CPIAUCSL",
              "Federal Expenditures":"W068RCQ027SBEA",
              "Federal Receipts":"FGRECPT",
              "Federal Debt": "FYGFDPUN",
              "$CPI_{Turkey}$":"TURCPIALLMINMEI",
              "Real GDP":"GDPC1",
              "GDP":"GDP",
              "Fed: Total Assets": "WALCL",
              "Fed: Treasuries":"TREAST",
              "Federal Deficit": "M318501Q027NBEA",
              "$M_{2_{Turkey}}$":"MYAGM2TRM189N"}

rate_codes  = {"$r_{UST_{30y}}$": "DGS30",
               "$r_{Mortgage_{30y}}$": "MORTGAGE30US",
               "$r_{Discount_{Turkey}}$":"INTDSRTRM193N"}

start = datetime.datetime(1960, 1, 1)
end = datetime.datetime.today()

FRED_dct = {"Level": gather_data(data_codes = data_codes, 
                                 start = start, 
                                 end = end, 
                                 freq = "M")}
rates = gather_data(data_codes = rate_codes, 
                                 start = start, 
                                 end = end, 
                                 freq = "M")
FRED_dct["Level"]["Fed Remittances to Treasury"] = web.DataReader("RESPPLLOPNWW", 
                                                                  "fred", 
                                                                  start, 
                                                                  end).resample("M").sum().div(1000).mul(12)

for y in ["Federal Expenditures","Federal Receipts", "Federal Debt", "GDP", "Real GDP"]:
    FRED_dct["Level"][y].interpolate(method='polynomial', order=3, inplace = True)  
rates["$\\frac{Federal Debt}{GDP}$"] = FRED_dct["Level"]["Federal Debt"].div(FRED_dct["Level"]["GDP"].mul(1000))

FRED_dct["Level"]["$\\frac{Deficit}{GDP}$"] = FRED_dct["Level"]["Federal Deficit"].div(FRED_dct["Level"]["GDP"])
FRED_dct["Level"]["$\\frac{FedAssets}{GDP}$"] = FRED_dct["Level"]["Fed: Total Assets"].div(FRED_dct["Level"]["GDP"].mul(1000))
FRED_dct["Level"]["$\\frac{FedTreasuries}{GDP}$"] = FRED_dct["Level"]["Fed: Treasuries"].div(FRED_dct["Level"]["GDP"].mul(1000))
FRED_dct["Level"]["$\\frac{Remittances}{Federal Receipts}$"] = FRED_dct["Level"]["Fed Remittances to Treasury"].div(
    FRED_dct["Level"]["Federal Receipts"])
FRED_dct["Log"] = np.log(FRED_dct["Level"])
FRED_dct["Rates"] = FRED_dct["Log"].diff(12)

for rate, val in rates.items():
    for key in FRED_dct.keys():
        FRED_dct[key][rate] = val
```

## Conceptual Application: Turkey

We can better understand the argument of Sargent and Wallace if we observe the dynamic between inflation and tightening in a country where tightening ultimately failed. Turkey has not had a record of success with regard to maintaining low and stable inflation. The reason for this is obvious given data from $M_2$. Monetary expansion appears to be responsible for thge rise in the price level. Notice that even during periods during which the rate of monetary growth reduced temporarilty (i.e., 1994), inflation did not fall. In fact, the upward surprise that followed the attempt at reduction only made inflation worse. 

```python
plot_vars = ["$r_{Discount_{Turkey}}$", "$CPI_{Turkey}$", "$M_{2_{Turkey}}$"]
plot_df = FRED_dct["Rates"][plot_vars]
fig, ax = plt.subplots(figsize = (20,12))
plot_df[["$CPI_{Turkey}$","$M_{2_{Turkey}}$"]].loc["1985":].plot.line(linewidth = 5,
                                                        ax = ax)
ax.set_title("Year-over-Year Rates")
ax.set_xlabel("")
```

Turkey has faced another devaluation in the last year. Yet, there is no $M_2$ data available for this period. We see the beginning of the struggle during 2018. Inflation began to pick up and the central bank, in response, increased the rate of interest. Perhaps since inflation had been relaibly lwo in previous years, investors were convinced by the sustained rate hike that this norm would continue. Even after inflation began to fall, the central bank maintained a high discount rate. However, the response of the centrual bank in 2021 was not enough to offset inflationary concerns. Investros viewed the high rate as insufficient and, likely, recognized that the higher rate might be followed by a lower rate. The rate of inflation persistntly rose, approaching 20\% year-over-year. When the central bank gave up this fight, investors appear to have been vindicated. Now Turkey faces an inflation rate of 60\%.

## Unconventional Monetary Policy and Fiscal Policy

The interaction between monetary and fiscal policy under QE, which we investigated last week, provides a similar sort of problem. The Federal Reserve has taken on an unprecenteded level of liabilities. When interest raets are low, this is not a problem. As we will see, rising interest rates have increased the value of regular payments made by the Fed to depositholders and those holding those lending to the Fed through its reverse repurchase agreements.

To begin, I want to emphasize that the Fed's response to the COVID-19 shutdown was relatively good. The key response, in my opinion, was the Fed's expansion of circulating currency. This was a very effective means of stimulating economic activity. As you can see, real GDP returned to trend relatively quickly. The Fed was slow to raise interest rate, however, even after real GDP had returned to trend. The Fed's response was about 6 months delayed, in which time its stimulus ceased to effectively raise real income. The recent roudn of tightening is both necessary and a signal to investors that the Fed is willing and able to tighten.

```python
plt.rcParams['axes.ymargin'] = 0.09
plt.rcParams['axes.xmargin'] = 0
plt.rcParams.update({"font.size":40})
axs = {}

for log_rate in ["Log","Rates"]:
    fig, axs[0] = plt.subplots(figsize = (20,12))
    axs[1] = axs[0].twinx()
    plot_vars = ["Real GDP","$r_{UST_{30y}}$"]
    for i, ax in axs.items():
        plot_var = plot_vars[i]
        color = "C" + str(i)
        FRED_dct[log_rate][plot_var].loc["2018":].dropna().plot.line(linewidth = 5,
                                                                      color = color,
                                                         ax = ax)
        ax.set_yticklabels([round(y,3) for y in ax.get_yticks()], color = color)
        ax.set_ylabel(plot_var, color = color)
    axs[1].set_yticklabels([str(y) + "%" for y in axs[1].get_yticks()], color = color)
```

This round of tightening, however, has new difficulties that have not been confronted previously. For example, the Federal Reserve's balance sheet is larger than it has ever been in terms of its size relative to GDP. As it winds down its balance sheet, the impact on interest rate in financial markets may be pronounced. However, we should also expect that as the Federal Reserve diminishes its influence over the level of resource allocation in financial markets, that this resource reallocation may result in some efficiency improvements as competitive market forces begin to exert greater influence over allocation.

However, the Fed needs to also make sure that tightening does not generate a liquidity crisis where rising rates on bonds diminish bank liquidity since U.S. Treasuries also serve as quasi-reserves due to Basel III Liquidity Coverage Ratio Requirements that force banks to hold a minimum proportion of High Quality Liquid Instruments for each unit of private, risky investment. Recently, Long-term rates have fallen in spite of Fed tightenign, which seems to indicate that 1) investors believe that the Fed's policy is effective and 2) the value of bank balance sheets will probably not be much further diminished by rising rates on U.S. Treausuries. At present, it seems the Fed has the leeway necessary in order to accomplish a successful tightening of policy.

```python
rename_dct = {
    "rgdpbarro":"$\\frac{y}{pop}$", 
    "iy":"$\\frac{I}{Y}$",
    "stir":"$i_{short}$",
    "ltrate":"$i_{long}$",
    "cpi":"$P_{CPI}$",
    "gdp":"Py",
    "narrowm":"$M_0$",
    "money":"$M_T$",
    "unemp":"U",
    "xrusd":"$P_M$",
    "revenue":"T",
    "expenditure":"G",
    "debtgdp":"$\\frac{G_{debt}}{y}$"}
plot_vars = list(rename_dct.values()) + ["crisisJST", "Annual Interest", "$\\frac{Interest}{T}$","$\\frac{T}{Interest}$"]

JST = pd.read_excel("https://www.macrohistory.net/app/download/9834512569/JSTdatasetR6.xlsx?t=1668516065",
             index_col = [2,0]).rename(columns = rename_dct )
JST["$G_{debt}$"] = JST["$\\frac{G_{debt}}{y}$"].mul(JST["Py"])
JST["Annual Interest"] = JST["$G_{debt}$"].mul(JST["$i_{long}$"].div(100))
JST["$\\frac{Interest}{T}$"] = JST["Annual Interest"].div(JST["T"])
JST["$\\frac{T}{Interest}$"]=JST["$\\frac{Interest}{T}$"].pow(-1)
data_dct={"Levels": JST[plot_vars],
         "Log":np.log(JST[plot_vars]),
         "Rates": np.log(JST[plot_vars]).groupby("iso").diff()}
rates = JST[["$i_{short}$","$i_{long}$"]]
rates["$\\pi$"] = JST["$P_{CPI}$"].groupby("iso").pct_change().mul(100)
for i in ("$i_{short}$","$i_{long}$"):
    rates[i.replace("i","r")] = rates[i].sub(rates["$\\pi$"])
for key in data_dct.keys():
    for r in rates.keys():
        data_dct[key][r] = rates[r]
plot_vars = plot_vars + ["$\\pi$"] + list(rates.keys())[-2:]
```
