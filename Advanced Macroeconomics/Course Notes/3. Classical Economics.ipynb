# Keynes and the *Classics*

The general view of the macroeconomy preceding Keynes was that markets promote harmony. Government intervention typically distrupts that harmony. Of course, the views of classical economists varied. The tendency to assert the ability of markets to promote harmony, however, harkens to Adam Smith who, as early as writing the _Theory of Moral Sentiments_ identified natural mechanisms that promote coordination in human society. This coordination was not, for Smith, purely economic. It drew fundamentally from sentiment in the human heart, which he asserted was prodded by the Impartial Spectator. Those who engaged in commerce, Smith observed, were most in accord with the Impartial Spectator as the success in commercial society was dependent upon the willingness of one's trading partners to cooperate with oneself. 

Typically, when John Maynard Keynes refers to Classical Economics, he is drawing from Say's formulation which asserted that, so long as prices adjust to changing conditions in the market place, which they do when not arbitrarily obstructed, the economy will tend toward full employment. We will discuss Say at greater length. For now, we will elaborate the conditions of full employment:


1. all economic agents (firms and households) are rational and aim to maximize their profits or utility; furthermore, they do not suffer from money illusion;
2. all markets are perfectly competitive, so that agents decide how much to buy and sell on the basis of a given set of prices which are perfectly flexible;
3. all agents have perfect knowledge of market conditions and prices before engaging in trade;
4. trade only takes place when market-clearing prices have been established in all markets, this being ensured by a fictional Walrasian auctioneer whose presence prevents false trading;
5. agents have stable expectations.

In the classical paradigm, as defined by Keynes, the economy exhibited a strong tendency toward the long-run outcome where nominal factors do not affect real factors. Another way of saying this is that money is neutral in the long-run. According to Say, there never exists a shortage of money, but rather, any apparent shortage simply reflected mispricing of consumption goods and productive factors. Interventions aimed at expanding the quantity of money owing to its dearness, then, were misplaced. While this description of Say is accurate, it is actually a misrepresentation of macroeconomics preceding Keynes, especially as it had developed immediately preceding Keynes's ascendance. For our immediate purpose, however, we will review production as framed within the classical paradigm as it was within this backdrop that modern macroeconomics was developed.


### Classical Theory of Employment and Output

As we saw with the Solow model, which is typically framed as the neoclassical view, the classical economists viewed the real sector as operating according to forces indpendent of monetary factors. The wealth of a society was dependent upon productive factors. It was from investment in capital and provision by capital owners of jobs for labor wherein that capital was made productive that wealth eminated. 

$$ Y = AF(K,L) $$

1. $Y =$ Real Output Per Period
2. $K =$ Quantity of Capital Used Per Period
3. $L =$ Quantity of Labor Inputs Used Per Period
4. $A =$ An Index of Total Factor Productivity
5. $F =$ Function Which Relates Real Outputs to Inputs $K$ and $L$

The dynamics of labor and capital markets are driven by the marginal product of each. With each additional unit of labor or of capital, how much new income is produced? The marginal production of labor must match or exceed the wage paid to labor, otherwise the firm will not earn a profit. Likewise, the marginal product of capital must exceed the cost of renting the capital, otherwise capital will not be worth renting.

$MPL = \frac{d Y}{d L}$

$MPK = \frac{d Y}{d K}$

### Diminishing Marginal Return to Capital and Labor

In the classical model of production, as well as in the Solow model which is a particular elaboration of the model, we presume diminishing returns to capital and labor. Each additional unit produces a bit less than what the previous unit employed had produced. As we saw with the Solow model, diminishing marginal returns to investment in capital and labor adjoined to constant marginal costs of employment ensure that a limit of employment is reached where MPK and MPL equal or are exceeded by rents and real wages, respectively.

### $\frac{d^2y}{dK^2} < 0$

### $\frac{d^2y}{dL^2} < 0$

$\pi_{max}:P_i \Delta Q_i = w_i \Delta L_i$

$\frac{\Delta Q_i}{\Delta L_i} = \frac{W_i}{P_i}$

#### In Equilibrium

$P_i= MC_i=\frac{W_i}{MPL_i} $


#### Supply of and Demand for Labor

Both the supply of and demand for labor are a function of real wages. All agents interpret the worth of their nominal income in light of the real goods that their income allows them to require. We will see later that inflation may temporarily lower real wages, thus increasing the quantity demanded for labor, however, only improvements in the productivity of labor can shift demand for labor to the right. 

Since we do not divide income, $y$, by labor, we can observe that an increase in $A$ draws labor into the market. This improvement in the effectiveness of capital improves the productive efficiency of capital, thereby improving the marginal product of labor. Of course, an improvement in efficiency, on its own, will have only limited affect on the quantity of labor employed. That improvement in efficiency must be allowed to improve the level of investment in capital, thereby increasing $K$. Each additional unit of capital improves demand for labor. As productivity improves and capital expands as a result, demand for labor and the average real wage of labor also improve.

This was not always the status quo. While growth did occur before the industrial revolution, the tendency was for an expanding labor force to promte **extensive growth**, where unused resources were brought into production. This may have been the result of modest technological development that enabled expansion, but until at least the 18th century, technological development did promote improvement in efficiency that would significantly raise the wage of labor. In fact, it was only as a result of a wave of deaths from the bubonic plague during the 14th century that wages rose leading into the Enlightenment.

In recent decades, technogological improvement has enable **intensive growth**, allowing for more efficient employement of resources already exploited. While the dividing line between technological improvement that enales either kind of growth may be fuzzy, much economic expansion over last century has clearly been of the intensive sort as innovation, not expansion of the labor force, has been precursor to economic development. For example, more efficient transportation and communication,  regularization and automation of productive processes have played a fundamental role in improving incomes.

#### Growing effective capital stock improves demand for labor
The demand for labor is a function of the quantity of capital. As the supply of capital increases, holding labor constant, so too does the return to employing a unit of labor. It is easy to demonstrate this by referring to Solow's effective capital unit, $\frac{K}{L}$, which controls for labor by including it in the denominator. Thus, in the presentation of the classical model by Snowdon and Vane, it makes sense to view an increase in demand for labor, over the long-run, as reflecting an increase in effective capital. 

It is important to remember that much, though not all, of the emphasis of classical economics was on the long-run. The long-run refers to a state where the operation of real factors is not impacted by adjustments in nominal factors - i.e., money and the price level. Over the short-run, the return to capital fluctuates, not from fundamental factors, but in light of expectations - optimism or a lack thereof, which then influence the direction of aggregate demand. Short-run fluctuations in the return to capital equivaliently impact the returns to labor, since the returns are being impacted by changes in the price of the output sold. Thus, demand for capital and labor owing to short-run variations operate in the same direction.

### Labor-leisure Tradeoff

As wages rise, more labor is incentivized to enter the work force. The classical model assumes that their is no labor leisure tradeoff, however, as worker productivity has improved over the last two centuries, their has been a noticeable decline in the average number of hours worked by employees. The obvious explanation for this phenomenon is the satiation of necessity. If workers must choose between working and being impoverished, the choice not to work is extremely costly. On the other hand, if wages are relatively high, one might choose not to work and, as a result, trade luxury for a modest, but still comfortably, existence. Leisure, though, might not be the most proper name for this tradeoff as one might simply pursue ventures that provide nonmonetary gain. For example, participation in community life involves reciprocatory relationships that require a high level of commitment. Such participation does not comprise leisure but, rather, the value created in the context of non-pecuniary relationships.

### Say's Law of markets

Returning to Say's Law, we identify the fundamental structure of macroeconomics viewed in terms of aggregates. Real income can be dividied into consumption and investment. Really, it is more accurate to divide the economy between consumption and non-consumption. We call that non-consumption *savings*. It so happens that savings tends to translate into investment, particularly in the modern, capitalistic economy where financial market operate with relative efficiency. 

Savings and investment are coordinated by the interest rate. The interest rate is both the cost of borrowing and the return for investing savings. While we know that a saver chooses between holding onto funds for the sake of liquidity, as Keynes will emphasize, the classical paradigm emphasized that $S=I$. Savers  invest funds by lending to those engaged in productive activity or by engaging in productive activity oneself. The saver is more greatly incentivized to offer savings to the borrowing entrepreneur the higher are rates paid to the saver. Rates themselves are determined by the return to capital. Presuming that the rate of return on capital investment is constant, borrowers will want to borrow less as interest rates increase. If there is a mismatch between the rate of return on capital and the interest rate, borrowers will borrow more in the case that the rate of return is higher than the interest rate and less in the case that the rate of return is lower than the interest rate.  Of course, interest rates may rise or fall in response to movement higher or lower of return to capital. In equilibrium, the rate of return to investment in capital matches the interest rate.  

Further, the saver may choose to consumer less

$$Y - C(r) = I(r)$$

$$I(r) = S(r)$$

$$\frac{dC}{dr} < 0$$

### Quantity Theory of Money

Notice that nowhere in this discussion has there been objections concerning about the influence of money on output. This is somewhat ironic since 1) banks collectively influence the total quantity of money by the setting of the reserve ratio, and 2) changes in demand for money influence both the price level and the interest rate. In the classical view, changes in the quantity of money move the price level. In the long-run, money supply would equal money demand, with the price level to offset descrepancies between the two. Thus, the classical economists, at least as portrayed by Keynes as followers of Say, presumed away the significance of short-run disturbance in the suppy of and demand for money. 

This presumption places the classical view as operating in the long-run. Often, this is represented by the quantity theory. The quantity theory is an instance of the equation of exchange with the particular assumptions that velocity and income are constant. 


#### Equation of Exchange

$$MV \equiv Py$$

$$M_d = kPy$$
$$M_s = M_d$$

We can write this in dynamic form by logging and taking the total derivative:

$$ln(MV)=ln(Py)$$

$$ln(M)+ln(V) = ln(P)+ln(y)$$

Next we take the total derivative. Since these are sums, the transformation is intuitive:

$$\frac{dln(M)}{dM}+\frac{dln(V)}{dV}\approx \frac{dln(P)}{dP}+\frac{dln(y)}{dy}$$

$$\frac{dM}{M}+\frac{dV}{V}\approx \frac{dP}{P}+\frac{dy}{y}$$

#### Quantity Theory

$$M\bar{V} = P\bar{y}$$

Since v and y are constant, in the dynamic form $\frac{dV}{V} = 0$ and $\frac{dy}{y}$. Thus:

$$\frac{dP}{P} = \frac{dM}{M}$$
$$\pi = \frac{dM}{M}$$

Using discrete time:

$$\pi = \frac{\Delta M}{M}$$

$$\pi = \% \Delta M$$

We should be sympathetic to concentration on the quantity theory by classical economists. Whether owing to currency devaluation by princes and kings or importation of gold from the new world, price inflation resulting from a lowering of the value of the monetary unit was the major monetary difficulty to be explained in premodern and early modern societies. Only as financial markets grew and industrialization became the norm did it become clear that oscillations in demand for money would play an important role in determing the value of total expenditures. Still, the classical economists should be defended. In particular, those monetary economists that Keynes lumped together with the classical economists did not hold naive views concerning the relationship between financial markets, demand for money, and interest rates. Before defending these theorists, however, we will need to explore the views presented by John Maynard Keynes.

### Keynes

#### Backdrop of the Great Depression

John Maynard Keynes is thought to have revolutized macroeconomics, setting the basis for it as a separate discipline. While this fanfare is more perception than truth, Keynes did play a role in attracting interest to the macroeconomy and did provide novel framing and emphasis, which we will discuss in the approaching section.

Keynes began penning his General Theory in as he observed the severity of the Great Depression. having just published his *Treatise on Money*, Keynes was in a strong position to analyze the most severe and persistent economic crisis that the modern capitalist world has ever encountered. We will discuss the Great Depression in detail in approaching weeks. For now, some generally features are sufficient for identifying context.

The Great Depression can be powerfully modeled as a series of negative aggregate demand shocks. These arose in the form of 1) tight monetary policy, by the Bank of France and the Federal Reserve in particular, and 2) banking collapses both in the U.S. Europe starting in 1931. These negative aggregate demand shocks are reflected by a fall in both real and nominal aggregate incomes. This occurred in the form of a combination of falling levels of real production as well as falling price levels.

$-\% \Delta M$ and $-\% \Delta V \rightarrow -\% \Delta P$ and $-\% \Delta y$

```python
#FRED.py
import pandas as pd
import pandas_datareader.data as web
import datetime
import time

def gather_data(data_codes, start, end = datetime.datetime.today(), freq = "A",
               delay = 1):
    i = 0
    # dct.items() calls key and value that key points to
    for key, val in data_codes.items():
        if i == 0:
            # Create dataframe for first variable, then rename column
            df = web.DataReader(val, "fred", start, end).resample(freq).mean()
            df.rename(columns = {val:key}, inplace = True) 
            i = None
        else:
            time.sleep(delay)
            # If dataframe already exists, add new column
            df[key] = web.DataReader(val, "fred", start, end).resample(freq).mean()

    return df
```

#### Keynesian Macroeconomics: Aggregate Demand Shortfalls and Underemployment of Resources 


John Maynard Keynes proposed that microeconomic equilibra may be disrupted by a mismatch between savings and investment caused by extreme oscillations in the velocity of money. Keynes asserted that the instability of velocity could lead to persistent shortfalls in real income, with real income falling *instead of* the price level. For Keynes, the economy does not persistently operate in the long-run but, rather, typically reflected short-run macroeconomic forces.

If this was all that Keynes claimed, he actually would not have been out of line with macroeconomic thoguht of his era. This view of short-run disruption caused by volatility in nominal factors was entirely consistent with Irving Fisher's *Money Illusion* and concern about demand for gold reserves expressed by Ralph Hawtrey and Gustav Cassel. In fact, Hawtrey penned the *Good and Bad Trade*, which explained how oscilliations in demand for money drive changes in interest rates and the level of investment. Like Keynes, these authors recognized that the non-neutrality of money ran the following course:

$$+\Delta M \rightarrow -\Delta r \rightarrow +\Delta I \rightarrow +\Delta Y, + \Delta L$$

In his presentation, however, Keynes ignores the long-run impact of monetary expansion, a key point of controversey since rising interest rates would alleviate the zero-lower-bound problem. Instead, Keynes believed that persistent pessimism could lead to persistently low levels of velocity that translated to persistently defficient aggregate demand.

Kenyes did show some sympathy for the classical monetarist position, though. Speaking of Hawtrey, he writes:

> There are very few writers on monetary subjects from whom one receives more stimulus and useful suggestion . . . and I think there are few writers on these subjects with whom I personally feel more fundamental sympathy and agreement. The paradox is that in spite of that, I nearly always disagree in detail with what he says! Yet truly and sincerely he is one of the writers who seems to me to be most nearly on the right track! 

> Keynes (1930, 86) in Glasner, "Hawtrey and Keynes"

Keynes asserted that the velocity of money is not only unstable but under cases of extreme depression, demand to hold newly created money was infinite! Therefore, central bank expansion of the money stock may have been necessary during a downturn, but expanion alone was not sufficient for lifting the level of total expenditures to move the macroeconomy toward its long-run equilirbium. 

As a result, even in the case of monetary expansion, Keynes claimed that there would persist widespread underemployment of resources, especially labor. Keyne's assertion pushed against the general theme of economic theory that had developed since Adam Smith. Smith had identified a natural order in competitive markets where resources that were not employed toward their most valued use would be bought up at discount by those who could make fuller use of those resources. This held true for both capital and labor, though Smith did recognize that there could exist positions where labor had limited bargaining power. The them for Keynes, however, was that the economy tended toward disorder, largely due to uncertainty. The less confident one felt about the future, the more likely were they to curtail investment.

Keynes provided a solution to this problem by suggeesting that governnment be given the charge of lifting the circulation of the money stock by engaging in deficit spending during recession. The development of Keynes's doctrine in his *General Theory of Employment, Interest and Money* was the theoretically significant program of economic planning to have been promoted to this point. While Keynes did not consider himself a socialist, he believed that both fiscal and monetary policy should operate to promote efficiency in financial markets that would prevent aggregated demand shortfalls from hamring production in other sectors.

A key feature of his theory was his use of the *propensity to consume*. Keynes suggested that if economic agents consume a constant fraction of their income, then a positive, exogenous income shock would be multiplied through the economy.

$$C = a + cY$$

$$Y = a + cY + I$$

$$Y = \frac{a + I}{1-c}$$
$$0<c\le 1$$

Keynesian Multiplier:
$$k = \frac{1}{1-c}$$

$$ \Delta Y = \Delta Ik$$

If investors were seized by uncertainty, then the additional investment required to move the economy to macroeconomic equilibrium could not come from financial markets or even from monetary policy. In other words, Keynes was concerned that in a suboptimal equilibrium, $\Delta I = 0$. Instead, Keynes suggested an increase in G, such that: We could rewrite the original equation:


$$Y = \frac{a + I + G}{1-c}$$


Notice that now:

$$\Delta Y = \Delta Ik + \Delta Gk$$

Since $\Delta I = 0$

$$ \Delta Y = \Delta Gk$$


#### Liquidity Preference, Interest Rates, and Liquidity Trap

Keynes further justified expansionary fiscal policy by highlighting concern about disequilibrum in financial markets. Recall that according to the Fisher Equation:

$$i = r + \pi_e$$

And that in equilbrium:

$$r = \frac{\Delta y}{\Delta K} = \frac{\Delta y}{\Delta I} $$

Investors will continue borrowing so long the rate of returned earned by investment capital is greater than the rate of interest. This is true whether we are considering real returns or nominal returns that reflect inflation. During a recession, the rate of inflation would fall, often below zero. This would create dislocation financial markets. Suppose that to reach the equilibrium rate of interest, the nominal rate charged in financial markets would need to be negative. Since investors will, under reasonable circumstances, refuse to lend theirm money for a negative return - meaning that lenders would pay borrowers to borrow money - the result would be a *de facto* diminishement of the supply of loanable funds. The problem could be solved by generating inflation, but Keynes did not believe monetary policy could accomplish this. Instead, the zero lower bound problem meant that government could borrow to offset the gap generated by the zero lower bound problem. With excess capacity not in use, government borrowing would lift incomes and increase the level of employment of labor without lifting interest rates. This offseting of the gap in the loanable funds market would simultaneoulsy move aggregate demand back to a value representing full employment, where real income matched the value of what we call long-run aggregate supply.

#### Rigid Nominal Wages and Involuntary Unemployment

Aside from financial markets, Keynes located the source of consistently sluggish demand as a function of sticky prices. Keynes claimed that nominal wages, in particular, were downwardly inflexible. If demand for a good falls, typically a mix of a fall in the goods price and a fall in the quantity result. However, to the extent that the price of the good does not fall, the the quantity of the good demand will fall even more sharply. The logic is both consistent with our intuition and the data.  Falls in the average real wage are rare and and limited. In fact, in the most recent recession, real wages rose significantly as most of those unemployed were low-skilled workers whose job had required physical presence. Keynes's claim about rigid nominal wages seems to have some merit. In the least, a fall in nominal wages is typically not enough to lead to a fall in the average hourly wage. Instead, the quantity of labor employed falls.

Whereas the central has been seen as the lender of last resort, Keynes's normative inferences led him to view the state as the employer of last resort. During down turn, the same government borrowing that offset the borrowing shortfall in the loanable funds market would offset an excess supply of labor that would persist so long as nominally rigid wages promoted in increase in real wages as deflation set in.

# Conclusion

While one may or may not agree with the Keynesian program, one cannot deny that Keynes's influence on the direction of macroeconomic discourse was unprecedented. He became a prominent voice for discontented economists skepitcal of theoretical idealization of competitive markets.               Likewise, he brought into focus concerns about the impact of velocity shocks on aggregate demand. Further, he challenged the view that velocity was stable and that financial markets were self-correcting. 

It was this framing that macroeconomic discourse would follow until present. As we will see, there was a monetarist counter revolution, yet it was largely shaped by discourse with Keynesian opposition. The same concerns about the stability of velocity or lackethereof will continue to reappear. And it is within this discourse that revolutions in empirical analysis and economic modeling take place. And, as with discourse between Keynes and his contemporaries, one side of discourse tends to express faith in market mechanisms that coordinate resources, while the other believes that those mechanisms fail, either objectively or from the point of view of justice. 

A look into the context of a nascent Keynesian economics illuminates not only Keynes's context, but the entire history of macroeconomics that followed.
