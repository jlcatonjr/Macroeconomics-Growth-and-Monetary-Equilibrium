# Entrepreneurship and the Economic Forces of Supply and Demand

Economics is the study human action subject to scarcity. It observes action in terms of means and ends. The agent perceives her environment and is able to imagine how she may acquire some end. We assume that the agent values any end that she seeks to realize.  Individuals rank ends according to the magnitude of desire for the end in light of cost of attaining the end. This valuation reflects the improved utility that the agent expects to receive.  Economists often make the general and reasonable claim that individuals maximize utility. This is a fancy way of saying that humans act in the manner that they believe is most valuable. They act according to their own beliefs and preferences.

Realization of a good is always accompanied by opportunity cost. The more one values a good, the higher a price he or she is willing to pay for it. The higher the price paid for a good, the less one is able to consume other goods. From the notion of opportunity cost, we will derive both the supply curve and demand curves. These reflect the facts that as the price of something increases, all else equal, individuals will demand less of the good and producers will provide more of the good. 

We will investigate the existence and creation of value. We will consider the role of the entrepreneur in creating value for others in society and the role of exchange in allowing for the improvement of the welfare of producers and consumers.

## Goods and Value

Agents act to attain goods. You may act to attain a candy bar: a physical good. You must assume that you are better off with the candy bar than without it. It is possible that the candy bar will be a link in a long chain of consumption that degrades your health, but in the moment of action, you choose the candy bar whose utility from consumption is realized in the short term over good health that you would receive in the long term if you abstained from these habits. Regardless of its effect, short-term or long-term, a good is a good because a person values it and is therefore willing to incur a cost to attain it.

A good can also be a function or, in other words, a service or a process. Some people will talk to strangers on the phone simply to alleviate their loneliness. They have paid for a good that exists only at the time of consumption. Others pay significant amounts of money to play video games and may even buy artifacts in the game that have no existence outside of the game. Others will pay for characters in the game that have been matured by another player. Some will even pay for a surrogate mother (Kuchar 2016). All of these are things that the individual in question prefers given available options and a finite income.

In economics we identify actors by their function. One such function is  consumption. Consumption is the final use of a good which generates value (utility) for the consumer. We refer to such goods as final goods.  Those engaged in consumption are consumers. 

If you eat a parfait for dessert, then you have engaged in consumption. Some forms of consumption are less obvious. If you watch your television tonight, you may not think that you are consuming. You are consuming on several margins when you watch television. You consume electricity. You also consume the hardware in the form of marginal wear and tear. This is why the resale value of your television is lower than the its original sale value. In the same manner, you consume hardware that provides content for viewing.

Every good sold commands a price. Fundamentally, a price represents all things given up in order to posses or realize a good. In economics, we concentrate on money prices. You pay a money price when you pay $3.00$ for a gallon of gas or $1.00$ for a bottle of purified water. If a consumer uses money to purchase one of these goods, that means that he prefers the good purchased to the money spent.

All goods in the economy are produced for the end of final consumption as it is in consumption where individuals derive value. Some goods do not provide utility directly from their use. We refer to these goods as intermediate goods. The truck that hauled coffee beans from the processing plant to the café where you purchased a mocha was a means to your final consumption. As was the labor that ground the beans, crafted your mocha, and processed the transaction. These intermediate goods fulfill valuable function in bringing coffee to market. It is only by contribution to the provision of final goods that capital and labor merit positive prices: rent in the case of capital and of wages in the case of labor.

## Opportunity Cost and Moneyed Accounting

We discussed opportunity cost last chapter. In economics, the word cost means something different than accounting cost. Economists refer to opportunity cost, which represents the value of the option that a person would have chosen had they not preferred to attain a particular good. Every night, you must choose what meal you will eat for dinner. If you choose to eat fried chicken and potatoes, odds are you will not also eat curried fish with rice. If you are considering between these two options at dinner time and we assume that each meal commands the same price, the opportunity cost of choosing one meal is that you will not eat the other. We assume that you value the meal chosen more than the meal that you chose to forgo.

The introduction of money prices allow us to consider the notion of opportunity cost with great clarity. We typically interpret opportunity cost in terms of value denominated in currency. This is straightforward in terms of understanding the profit function. The entrepreneur incurs costs in the course of business. 

Clara, an independent web designer may pay for advertising, rent for an office, and for the equipment and software she uses to build websites. These are explicit costs. We can use explicit costs to calculate accounting profit. Clara's accounting profit each month is calculated as the difference between her revenues and explicit costs each month.

Opportunity cost includes explicit and implicit costs. It is possible that Clara earns an accounting profit but is actually incurring an economic loss. For example, she could design websites as an employee of a company. The wage that she would receive as an employee is included in her opportunity cost. Clara's economic profit includes both explicit and implicit costs. It is the difference between her revenues and this broad definition of costs. Economic analysis always considers opportunity costs. 

If Clara appeared to be incurring economic loss, we must assume that she values her work as a business owner in addition to the revenues she earns. If Clara decided to go back to being an employee because she felt that suffering the loss was not worthwhile, then this is evidence that she was truly suffering a loss in utility.

## Price and the Equimarginal Principle

If all action can be described by attempts to create the greatest value for the actor, then this means that an individual will engage in a behavior if the value gained compared to cost is relatively greater than for any other activity. In terms of ownership, we describe this in terms of utility gained per currency unit. Suppose that you you gain 10 units of utility from consumption of a candy bar priced at $2. Suppose that you could by a pack of gum for $1 from which you expect to gain 7 units of utility. In the first case you gain 5 units of utility per dollar spent. In the second case, you gain 7 units of utility in exchange for a dollar. 

Formally, this principle is described as the utility gained for each dollar spent on a good. A person will purchase more of a good if it is above the utility gained from any other good. In equilibrium, utility gained from possession of one good offers the same utility as utility gained from another.

$$\frac{MU_i}{P_i}=\frac{MU_{i+1}}{P_{i+1}}=. . . =\frac{MU_{j-1}}{P_{j-1}} = \frac{MU_j}{P_j}$$

The marginal utility gained for each dollar spent a good $i$ is equal to a dollar spent on any other good $i + n$ for all integers $n<j$. 

While this may be difficult to imagine for an individual, it is easier to think about this in terms of profit earned in different industries. Entrepreneurs will tend to be attracted to industries that offer relatively high rates of profit. The accumulate in these industries until the risk-adjusted rate of return for each dollar invested matches the rate of return in other industries.

## Supply and Demand

We analyze economic goods in regard to the benefit received by the consumer and cost incurred in it's production. A consumer who purchases a good believes that he will be better off with the good than if he kept the money that he spends in exchange for the good. We use this logic to construct schedules of supply and demand. A supply curve represents combinations of prices and quantities of a good produced by a single firm, a group of firms, or all firms in a market. As the price of the good rises, suppliers provide more of that good. A demand curve represents the quantities demanded for a particular good at different prices from an individual, a group, or for all consumers. As the price of a good falls, consumers demand more of that good.

Both supply and demand curves are a manifestation of marginal analysis. All action is a result of decision-making at the margin. We will assume that the value of one unit of a good is not dependent upon possession of another good. In some cases, the value attributed to consumption or possession of one unit of a good may be dependent upon the possession of other goods. A set of chess pieces that is missing one piece, say the black rook, gains significantly more aesthetic and functional value by the possession of that final piece. In most cases - those we will be analyzing in this book - value attributed to the possession of an additional unit of a good is lower than the value attributed to possession of the previous unit.

### Demand

Imagine that you are hungry. If you had not eaten for a day, you would greatly value a burger to sate your hunger. 
How much would you be willing to pay for this burger? Perhaps you would be willing to pay $5.00$. After you have eaten your first burger, you are less hungry. You would like to eat another burger, but you would not be willing to pay $5.00$ for the next one. Instead, you are only willing to pay $2.00$ for the second burger. After the second burger, you are not very hungry, however you do enjoy the taste of the burger, you would be willing to eat another for $0.50$. The fall in the price that you would be willing to pay for an additional burger represents a fall in the value that you attribute to each additional burger. We refer to this as the law of diminishing marginal utility.

Once you recognize the law of diminishing marginal utility, you will likely observe it in many areas of your life. Consider the decision to purchase a thumb drive to transfer files between to computers. If you have 4 gigabytes of data to transfer, we can assume that you will positively value a 16 gigabyte thumb drive. If all files fit on this drive, however, we can expect that a second drive of the same size will be less valued by you. What if you can only attain thumb drives that are one gigabyte in size? We can expect that you will load files that you most value – i.e., that you most need – on the first thumb drive. As you attain additional drives, you will transfer files that you value less or, in other words, that you less urgently need. Therefore, you will value each additional drive you acquire less than the previous drive. This lower valuation means that your willingness to pay for an additional drive falls with each additional drive you acquire.

In addition to the law of diminishing marginal returns, demand curves must slope downward out of economic necessity. The more expensive a good becomes relative to other goods, the more income the good consumes and, therefore, the less of the good you can buy. Likewise, the cheaper a good is relative to other goods, the greater is the income of a person who purchases the good. This is called the income effect. Suppose that the economy only had one good that you can purchase. If your income is $10.00$ and the good costs ten dollars, you can only purchase one of that good. If that good costs $5.00$, then you may buy two. The number will increase as the price of the good falls. Together, these principles strengthen our assumption that demand curves slope downward

#### Budget Constraint

Individual preference for goods comprise an ordering. At any moment, you necessarily prefer some goods over other goods. When you receive your pay each week, you must choose which food to buy and which not to buy. Over the course of a year you must purchase clothes. You must choose a means of transportation to facilitate other actions. Ludwig von Mises argues that these choices evidence a preference ordering. Action evidences preferences. 

Money spent on one good is money that cannot be spent on another good. In every situation, individuals are limited by a finite budget, whether that budget consists of wealth or time.  In Figure 2, we  represent Harold's choice between purchasing burgers or savings his money for other goods. If an individual chooses to purchase a burger whose price is $0.75$, that individual will be unable to purchase $0.75$ worth of any other good. Thus, every purchase comes at the cost of income that cannot be spent on other goods. 

Below we consider the evolution of Harold's expressed preferences as the price of burgers change. We assume Harold's preference function to be fixed, meaning that his preference for burgers over other goods will change only if the price of a burger changes or if Harold's income changes (Stigler and Becker, 1977). Thus, at any instant, Harold's preference for burgers against other goods can be fully represented before he makes his decision. In reality, we do not observe utility functions directly, but instead assume that Harold's actions evidence his preference function.

We shall assume that Harold's income is fixed at $12.00$ The quantity demanded by Harold will change only as the price changes. If a burger costs $0.25$, Harold will purchase $24$ burgers, if a burger costs $0.50$, he will purchase $12$ burgers, and if a burger costs $0.75$, he will not purchase any burgers.  We represent this data by showing three different budget lines. Each price for burgers represents a unique set of consumption bundles that can include burgers. Each time the price of a burger, PB, falls, Harold's potential income in terms of burgers increases. If Harold decided to spend the entirety of his income on burgers, he would buy $16$ burgers when $p_B = 0.75$, 24 burgers when $p_B = 0.50$, and 48 burgers when $p_B = 0.25$.

###Individual Demand Curve

We can use the data from Figure 1 to construct Harold's demand curve for burgers (Figure 2). To do so, we plot the pairs of price and quantity from Figure 1 on  a single plot and draw a line through these points. In the case where the points do not form a line, we would use a regression to estimate a line or curve through the points.

Recall the general linear equation:

$$ y = mx + b $$

We replace $y$ with $p$ and $x$ with $q$:

$$p=mq+b$$

To construct the line, we use the slope equation:

$$m = \frac{p_2 - p_1}{q_2 - q_1} $$

For example:

$$q_1, p_1 = 24, .25$$

$$q_2, p_2 = 12, .5$$

$$m=\frac{0.5 -0.25}{12 - 24}$$

$$ m = -\frac{.25}{12} $$

$$m=-\frac{1}{48}$$

Thus:

$$p = -\frac{1}{48}q+c$$
$$p = -\frac{q}{48}+c$$

We know also that our line crosses the vertical axis at PB = $0.75. Since c represents the value of the intersection between the line and the vertical axis, we write the full equation:

$$ p = -\frac{q}{48}+0.75 $$

## Market Demand Curve

The market demand curve represents the combination of all individual demand curves. Imagine that there are 100 people with exactly the same preferences as Harold. We sum their preference across the demand curves to represent the market demand for burgers. Since we are aggregating the quantity demanded at each price, we must first rewrite the demand function so that quantity is a function of price.

$$ p = -\frac{Q}{48}+0.75 $$

$$ p - 0.75 = -\frac{Q}{48}$$

$$ Q = -48p_B+36 $$

Now we may aggregate across consumer preferences. The burger market includes 100 people, so we may simply multiply each value by 100.

$$Q_B = -4800p_B+3600 $$

We restore the original structure, rearranging the equation once more.

$$Q_B - 3600 = -4800 p_B$$

$$ -\frac{Q_B - 3600}{4800}=p_B $$

$$ p_B=-\frac{Q_B}{4800}+0.75 $$

## Shifting the Demand Curve
We represent individual preferences at a moment in time using the demand curve. It is possible that  preferences change or that incomes grow or shrink. Either of these will shift a demand curve. While many transformations are possible, we generalize these and refer to increases or decreases in demand. An increase in demand occurs when the quantity demanded of a good at each price increases. We refer to this as a rightward shift in demand. A decrease in demand occurs when the quantity demanded of a good at each price decreases. We refer to this as a leftward shift in demand.

### Increase in Demand
Suppose that there is an increase in the preference of individuals to eat beef as opposed to chicken. Since burgers are made of beef, we represent this change in preference as a rightward shift in demand for burgers. Algebraically, we denotes this as an increase in quantity demanded at a given price. We shall hold the slope of the line constant and increase the Q-intercept of our equation, thus yielding:

$$Q_B = -4800p_B + c$$

Suppose that now consumers demand $2400$ burgerts at a price of $0.60$. We use these values to solve for the new Q-intercept:

$$2400=-4800(0.60)+c$$

$$2400=-2880+c$$

$$ c = 5280 $$

The new equation for demand is:

$$Q_B=-4800p_b+5280$$

We rewrite this with the price on the left hand side:

$$4800p_B=-Q_B+5280$$

$$p_B=-\frac{Q_B + 5280}{4800}$$

$$p_b = \frac{Q_B}{4800}+1.1 $$

The new P-interecept is $1.10$ with slope remaining the original value of $\frac{1}{4800}$.

### Supply

As with demand curves, the features of supply curves are generated by the assumption that individuals prefer to act in a manner that will bring the greatest value. A person will not be willing to incur a cost if he or she does not expect that the benefit of a certain action exceed this cost. This means that individuals will give up resources that they value least in order in order to produce goods that are of greater value. As was noted before, the difference between the value generated and the cost of creating the good represents the profit of the entrepreneur.
If entrepreneurs first give up resources that they value least in order to produce a good, this means that the value of resources used in production increases as the quantity of goods produced increases. Thus, the quantity supplied increases as the price increases. Supply curves slope upward.

## The Law of One Price

We will see that markets promote the correct price and quantity as defined by supply and demand curves. The market directs production by movement of prices that reflect changes in demand and in the cost of supply. Actors in the market are incentivized to promote this change by the profit motive. If there are gains from trade potentiated by discrepancies between prices of like goods, these discrepancies will be offset. We call the act of exploiting gains that arise from price discrepancies arbitrage. 
If the price of an ounce of gold in England is $1500 and the price of an ounce of gold the U.S. is $1505, there exist an opportunity to profit by buying gold from England and selling it in the United States. If the cost of buying an ounce of gold in the U.S. and selling it in England is less than five dollars, we expect that the final price in either country will lie somewhere between $1500 and $1505. If the cost of buying and selling and ounce of gold is zero dollars, then we expect that both countries will reach exactly the same price. 
There are cases where the law of one price appears to be violated, however, this is always due either to transaction costs or subjective valuations of acquiring a good. Consider two apartments that are identical in every way except location. If the apartment in New York is worth four times the value of the price in Omaha, it is not because the law of one price has been violated. The goods are commonly referred to as being non-tradeable. One good cannot be used for the same purpose of another. Not only are transportation costs especially high for real estate, the conditions that exist in one area, including scarcity, cannot be transported to another area. 
There also exists discrepancies between observed prices of tradeable goods. A gas station on one street corner may charge a higher price than one on the opposite corner and still not lose customers. Often, this may occur because the price discrepancy is small and drivers are willing to pay this cost in order to stay in the comfortable zone defined by their habits. But if the price of gas at one station increases enough relative to the price of gas at the other station, drivers may choose to purchase gas at the station across the street. 
The profit mechanism, driven by the desire of human agents to improve the state of the world that they drives the market process. The economic system promotes a high level of coordination where plans of economic actors exhibit a long-run tendency to cohere with one another. 

## Equilibrium

Thus far, we have simply selected prices to show how the quantity supplied and the quantity demanded adjust to changes in prices. However, prices are not chosen arbitrarily in the market place. They are determined by the joint forces of supply and demand. Competition in the market place bring the quantity supplied and quantity demanded of a good to match one another.

This can be illustrated by the supply and demand curves. Recall the equations that define demand and supply:

$$ Demand: p_B=-\frac{Q_B}{4800}+0.75 $$

$$ Supply: p_b = \frac{Q_B}{4800} - 0.25 $$

We can rewrite this to solve first for the equilibrium quantity:

$$ -\frac{Q_B}{4800} + 0.75 = \frac{Q_B}{4800}-0.25 $$

$$\frac{Q_B}{2400}=1 $$

$$ Q_B = 2400 $$

We can use either the demand or supply equation to yield the equilibrium quantity value. We will use the supply equation:

$$ p_B = \frac{2400}{4800} - 0.25 $$

$$ p_B = 0.5 - 0.25 $$

$$ p_B = 0.25 $$

The equilibrium quantity of $2400$ burgers is reached when the price of a burger is $0.25$. At a price that is any higher, the quantity supplied will be greater than the quantity demanded. At a lower price, the quantity supplied will be less than the quantity demanded.

## Price Dynamics and Shifts in Demand and Supply

A shift in one curve results in a change of price and quantity provided in the market. There are only two types of shifts. Leftward shifts represent a decrease in the curve shifted. Rightward shifts represent an increase in the curve shifted. Depending on the reason, a shift of a curve in one market can also have effects on curves in other markets.

We will introduce the different effects of possible shifts and follow with possible reasons and effects of shifts in each curve.

### Shifts in Demand

A shift in demand is positively correlated with the change in price and quantity that results. An increase in demand for a good will have a positive effect on price and the quantity of the good demanded while a decrease in demand will have a negative effect on price and the quantity of the good demanded.

### Changes in Income

A change in demand can be driven by a change in preferences, which we take as given, or they can be an effect of a change in income. The effect of a change in income on demand is not obvious. An increase in income could increase or decrease the demand for a given good. For most goods, an increase in income will increase demand for the good. Normal goods are goods whose demand is positively correlated with changes in income. As you grow more wealthy, it is likely that you will purchase more computer technology or at least a greater quantity of processing power in terms of random access memory (RAM), processing speed in terms of aggregate processing speed and number of cores, and hard drive space. Thus, computer technology is a normal good. Some other normal goods include air travel, food services, and housing in terms of square footage.

Goods whose demand is negatively correlated with changes in income are called inferior goods. As a college student, you may eat a significant amount of ramen noodles and pizza that you buy frozen from the store. Odds are, you will not tend to eat more of these when you graduate and begin to earn a higher level of income. Thus, frozen pizza and ramen noodles are an inferior good. Our discussion generalizes across the population. It is, of course, possible that an individual consumes more frozen pizza and ramen noodles, in which case these goods would be normal goods for the individual in question.

### Changes in the Price of Substitutes 

It is possible that demand for a good changes as the results of a change in the price of a related good. Related goods are either substitutes or complements. A substitute for a given good is one that can be used in place of that good. For example, consider the effect of an increase in the supply of almond butter on the demand for peanut butter. An increase in the supply of almond butter will cause the price of almond butter to fall relative to the price of peanut butter. Some consumers who view the two as substitutes will being to consume more almond butter and less peanut butter. The greater the magnitude of the fall in the price of almond butter relative to the price of peanut butter, the greater will be the fall in demand for peanut butter that results. The magnitude of the fall in demand for peanut butter is dependent upon the strength of consumer preference for almond butter over peanut butter.



> <img src="https://github.com/jlcatonjr/Macroeconomics-Growth-and-Monetary-Equilibrium/blob/main/Chapter%202/Figure%2016.png?raw=true" alt=" Supply of Almond Butter Decreases" title="Supply of Almond Butter Decreases"/>
> <center> <b>Figure 16

Suppose that  new consumers enter the market for almond butter who had previously not eaten either almond butter or peanut butter. The additional consumption of almond butter will drive up demand for almond butter and, therefore, drive up price. Some consumers who ate almond butter before the increase in demand from new consumers will substitute away from almond butter and toward peanut butter. Thus, an increase in demand for a good from a new base of consumers will also increase the demand for substitutes. Likewise, a fall in demand for a good will lead to a fall in demand for its substitutes.

> <img src="https://github.com/jlcatonjr/Macroeconomics-Growth-and-Monetary-Equilibrium/blob/main/Chapter%202/Figure%2017.png?raw=true" alt="Demand for Almond Butter Increases" title="Demand for Almond Butter Increases"/>
> <center> <b>Figure 17

We must be careful in our evaluation of a change in a market. Suppose demand for almond butter increases, but this occurs due to an autonomous change in consumer preferences. Consumers of peanut butter may become more conscious of their health. If they believe almond butter is healthier than peanut butter, then they substitute almond butter for peanut butter. Demand for almond butter has increased, but in this case it is associated with a fall in the price of peanut butter. 

The substitution effect still operates in this situation as a second order effect. Those who had previously consumed almond butter and who are willing to substitute peanut butter for almond butter will leave the market for almond butter. Thus the fall in demand for peanut butter that was initiated by an increase in health consciousness of some consumers is partly offset by consumers of almond butter who are willing to substitute to peanut butter as the price of almond butter rises.

#### Changes in the Price of Complements
As with substitutes, changes in the prices of complements demand for a compliment can be ambiguous. Suppose that demand for almond butter increases, but that this increase in demand for almond butter is not from consumers who prefer to eat almond butter with honey. Those who were eating almond butter and honey before the price change may find themselves eating less of the dessert as the price of this bundle has increased. Thus, the increase in the price of almond butter leads to a decrease in demand for honey. Likewise, a decrease in the price of almond butter will lead to an increase in demand for honey.

## Changes in Technology and the Effect on Supply

The primary driver of changes in the supply curve is change in technology. We define technology broadly. It includes both the structure of physical goods (like computers) as well as social organization. In the latter case, democracy and its manifestations are a kind of technology. Changes to either kind of technology effect the efficiency of capital.
In the modern market, it is difficult to imagine a scenario where a change in physical technology decreases efficiency. We live in an age of improving technology. However, we are not guaranteed that our social technology continues to improve the efficiency of production. Such technology includes systems of governance, whether formal or informal, and expectation that we form of ourselves and one another. The development of a Bourgeois Dignity in the Western Europe during the 17th and 18th centuries, for example, integrated the merchant class into English governance. Merchants were allowed more freedom and influence in society largely due to a growing belief that market activity was, in fact, honorable. The result was an explosion of economic growth whose effects ripple forcefully into the present. A belief of the opposite sort, as occurred in China under the early rule of the Chinese Communist Party, can be devastating. Many millions died as private property lost sanction with the rise of the communist party. Since, we have seen a transformation of institutions in China that have proved valuable in increasing the efficiency of production and exchange, thereby allowing entrepreneurs in the country to generate wealth.
So long as institutions secure ownership of property and the rule of law, we will continue to observe a dynamic in markets where entrepreneurs invest to increase the efficiency of production for goods most highly demanded. As population has increased over the last several centuries, so too has demand for food. Innovations in transportation and agriculture have operated to greatly reduce the cost of food. The general process of economic growth is illustrated in Figure 16. In the case of food, demand for food has grown. Entrepreneurs have invested in technology in response. Innovations in production and transportation of food have greatly reduced the cost of supplying food to consumers, thereby increasing the quantity provided and providing it at a lowering price.

## Price Controls

Profit seeking entrepreneurs and cost minimizing consumers tend to drive prices and quantities of goods produced toward the equilibrium values determined by supply costs and consumer preferences. It is possible, however, that governing authorities attempt to dictate price.

> <img src="https://github.com/jlcatonjr/Macroeconomics-Growth-and-Monetary-Equilibrium/blob/main/Chapter%202/Figure%2019.png?raw=true" alt="Price Floor" title=" Supply of Almond Butter Increases"/>
> <center> <b>Figure 19


Authorities can implement minimum or maximum prices. Minimum prices are called price floors, since it is illegal to sell below the dictated price. A price floor that is set above the equilibrium price that would arise without the intervention will lead producers to create more of a good than will be demanded by consumers since they receive greater revenue for each good sold. In the past, governments have purchased the excess good that are produced due to a price floor. It is not uncommon for the government to issue subsidies in order to create the consequences intended by governing authorities.

A common price floor is the minimum wage. It dictates the value of the monetary wage received by workers. The minimum wage limits the number of workers that an employer can profitably hire. If employers pay workers according to their productivity, then those who marginal workers who were employed before the implementation of a minimum wage will find themselves no longer employed.


> <img src="https://github.com/jlcatonjr/Macroeconomics-Growth-and-Monetary-Equilibrium/blob/main/Chapter%202/Figure%2020.png?raw=true" alt="Price Ceiling" title="Price Ceiling"/>
> <center> <b>Figure 20

Maximum prices are called price ceilings, since prices cannot be raised above the price ceiling. When a price ceiling is set below the equilibrium price, consumers will demand more of the good than producers are willing to provide. To produce these extra units at the price set by the price floor would generate losses for producers.

A common price ceiling is a rent control. Rent control that are priced below the market price lead to a number of unintended consequences. Firstly, they lead to an under supply of housing relative to demand. Landlords may attempt to offset the rent control by deriving value elsewhere. They demand large deposits that will not be returned to the renter. They may also indulge their own preferences when choosing who to allow to rent from them. For example, a landlord who is distrusting of those who don't share his or her culture may, on the margin, allow this preference to impact his choice of renter.

# Conclusion

Economics studies human action as defined by ends and means. From this framing, we develop principles to guide analysis. Individuals choose to act in a manner that they believe will generate the most value. When choosing among like goods, they choose the lowest cost option. When choosing among goods of the same cost, they choose the good that they value most. When the rule of law is present and when outside intervention is absent, economic exchange leads to relatively high levels of welfare for those involved, since each party expects that he or she is made better off by the exchange.
